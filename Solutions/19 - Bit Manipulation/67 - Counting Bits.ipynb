## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I solve this in 3 different ways.

1. First Approach: straightforward by using built-in functions.
  -

  ```python
  def countBits(n):
        L = []
        for i in range(n+1):
            bi = list(bin(i)[2:])
            L.append(bi.count("1"))
        return L
  ```


2. Second Approach: We can optimize the countBits function to achieve a linear time complexity and a single pass without using any built-in functions. One efficient approach is the [Brian Kernighan's Algorithm](https://www.educative.io/answers/what-is-kernighans-algorithm) for counting set bits in an integer.
  -

  ```python
  def countBits(n):
        counts = [0] * (n + 1)  # Initialize an array to store bit counts

    for i in range(1, n + 1):
        num = i
        count = 0
        while num:
            num &= (num - 1)  # Clear the rightmost set bit
            count += 1
        counts[i] = count
    return counts
  ```

3. Third Approach: there is another efficient way to count the number of set bits in an integer using a technique called "Lookup Table." This approach also allows us to count the bits in constant time for each integer.
  -

    ```python
  def countBits(n):
        counts = [0] * (n + 1)  # Initialize an array to store bit counts

     for i in range(1, n + 1):
        counts[i] = counts[i >> 1] + (i & 1)
     return counts
    ```

## 5. Code Implementation <a name="Code-Implementation"></a>


````python
class Solution(object):
    def countBits(self, n):
        counts = [0] * (n + 1)  # Initialize an array to store bit counts
        for i in range(1, n + 1):
            counts[i] = counts[i >> 1] + (i & 1)
        return counts
````

````python
#Example 1
n = 2
result = Solution().countBits(n)
print(result)
````

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
The code iterates from 1 to `n`, performing constant-time operations for each integer. The operations involve bit shifting (`i >> 1`), bitwise AND (`i & 1`), and array access (`counts[i >> 1]`). All these operations take constant time regardless of the size of `n`.

Therefore, the time complexity of the code is $O(n)$, as we process each number from 1 to `n` in a single pass.

Space Complexity:
The code uses an additional array, `counts`, to store the bit counts for each integer. The size of the `counts` array is `n + 1`, as we need to store the counts for numbers from 0 to `n`.

Hence, the space complexity of the code is $O(n)$ since the space required by the `counts` array grows linearly with the input `n`.
