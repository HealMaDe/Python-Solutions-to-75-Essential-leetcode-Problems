## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I solve this in 2 different ways.

1. First Approach: straightforward by using built-in functions. But this is not efficient in terms of time and spcae complexity
  -

  ```python
  def singleNumber(nums):
        n = len(nums)-1
        while n!=1:
            a = nums.pop(n)
            if a in nums:
                nums.remove(a)
                n -= 2
            else:
                return a
  ```


2. Second Approach: To solve this problem with a linear runtime complexity and using only constant extra space, we can utilize the bitwise XOR operation. XORing a number with itself results in 0, so if we XOR all the numbers in the array, the duplicate numbers will cancel out, and the remaining XOR result will be the unique number.
  -

  ```python
  def singleNumber(nums):
        result = 0
        for num in nums:
            result ^= num
        return result
  ```



## 5. Code Implementation <a name="Code-Implementation"></a>


````python
class Solution(object):
    def singleNumber(self, nums):
        result = 0
        for num in nums:
            result ^= num
        return result
````

````python
#Example 1
nums = [2,2,1]
result = Solution().singleNumber(nums)
print(result)
````

````python
#Example 2
nums = [4,1,2,1,2]
result = Solution().singleNumber(nums)
print(result)
````

````python
#Example 3
nums = [1]
result = Solution().singleNumber(nums)
print(result)
````

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:
The function iterates over each element in the `nums` array once and performs the XOR operation for each element. Since this operation takes constant time, the time complexity of the function is $O(n)$, where $n$ is the length of the `nums` array.

Space Complexity:
The function uses only a constant amount of extra space. It initializes a single variable `result` to store the XOR result. Regardless of the size of the input array, the space used remains the same. Therefore, the space complexity of the function is $O(1)$.
