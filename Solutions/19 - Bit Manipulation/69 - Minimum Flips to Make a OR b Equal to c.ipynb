## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use bitwise operations and bit shifting to process the bits of a, b, and c from right to left. We compare the bits and increment the flips counter accordingly.

1. Iterate through the bits of the given numbers a, b, and c
2. Count the minimum number of flips required to make the bitwise OR of a and b equal to c.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def minFlips(self, a, b, c):
        flips = 0
        while a or b or c:
            bit_a = a & 1
            bit_b = b & 1
            bit_c = c & 1

            if (bit_a | bit_b) != bit_c:
                if bit_c == 1:
                    flips += 1
                elif bit_a == 1 and bit_b == 1:
                    flips += 2
                else:
                    flips += 1
            a >>= 1
            b >>= 1
            c >>= 1

        return flips
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:
The method iterates through the bits of the numbers `a`, `b`, and `c` until all bits have been processed. In the worst case, the number of iterations is determined by the number with the most significant bit, which has a value of $log_2(max(a, b, c))$. Therefore, the time complexity of the method is $O(log(max(a, b, c)))$.

Space Complexity:
The method uses a constant amount of extra space, regardless of the input size. It only requires a few additional variables to store the current bits of `a`, `b`, and `c`, as well as the `flips` variable to keep track of the minimum flips. Therefore, the space complexity of the method is $O(1)$.
