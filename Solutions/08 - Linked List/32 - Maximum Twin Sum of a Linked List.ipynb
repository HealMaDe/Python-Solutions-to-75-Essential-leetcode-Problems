## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

1. iterate through the linked list
2. store the values of the nodes in an array
3. calculate the twin sum for each pair of nodes
4. find the maximum twin sum.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def pairSum(self, head):
        values = []
        node = head
        while node:
            values.append(node.val)
            node = node.next

        max_twin_sum = 0
        n = len(values)
        for i in range(n // 2):
            twin_sum = values[i] + values[n - 1 - i]
            max_twin_sum = max(max_twin_sum, twin_sum)

        return max_twin_sum
```

```python
#Example 1
head = ListNode(5, ListNode(4, ListNode(2, ListNode(1))))
result = Solution().pairSum(head)
print(result)
```

```python
#Example 2
head = ListNode(4, ListNode(2, ListNode(2, ListNode(3))))
result = Solution().pairSum(head)
print(result)
```

```python
#Example 3
head = ListNode(1, ListNode(100000))
result = Solution().pairSum(head)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Iterating through the linked list to store the values in the array takes $O(n)$ time, where $n$ is the number of nodes in the linked list.
- Calculating the twin sum for each pair of nodes takes constant time since we are accessing the values directly from the array.
- Therefore, the overall time complexity of the solution is $O(n)$.

Space Complexity:
- We use an additional array to store the values of the nodes. The size of the array is equal to the number of nodes in the linked list.
- Therefore, the space complexity of the solution is $O(n)$, where $n$ is the number of nodes in the linked list.

Overall, the solution has a linear time complexity of $O(n)$ and a linear space complexity of $O(n)$, where $n$ is the number of nodes in the linked list.
