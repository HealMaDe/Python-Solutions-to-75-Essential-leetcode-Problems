## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I successfully implement both approaches mentioned in the question for reversing a linked list:

1. **Iterative Approach:**
   - The iterative approach uses two pointers, `prev` and `current`.
   - It traverses the linked list, reversing the direction of the pointers as it goes.
   - At each step, the `current.next` is set to point to the `prev` node, effectively reversing the link.
   - This process continues until the end of the linked list, and the `prev` pointer becomes the new head.

2. **Recursive Approach:**
   - The recursive approach uses a recursive function to reverse the linked list.
   - The base case is when the current node is `None` or the last node.
   - The function recursively reverses the remaining portion of the linked list.
   - After the recursive call, it updates the `next` pointer of the current node to point to itself and sets the `next` pointer to `None`.
   - This process is repeated until the entire linked list is reversed.

Both approaches achieve the same resultreversing the linked list. The iterative approach is often preferred in practice for its simplicity and constant space complexity, while the recursive approach is more elegant but uses additional space due to the function call stack.



## 5. Code Implementation <a name="Code-Implementation"></a>
Iterative Solution:

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head):
        prev = None
        current = head

        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        return prev
```

Recursive Solution:

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head):
        if not head or not head.next:
            return head

        new_head = self.reverseList(head.next)
        head.next.next = head
        head.next = None

        return new_head
```

```python
#Example 1
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
result = Solution().reverseList(head)
print(result)
```

```python
#Example 2
head = ListNode(1, ListNode(2))
result = Solution().reverseList(head)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png



### Iterative Approach:

**Time Complexity:**
- The iterative approach has a time complexity of $O(n)$, where $n$ is the number of nodes in the linked list.
- This is because each node is visited once during the traversal.

**Space Complexity:**
- The space complexity is $O(1)$ as it uses a constant amount of extra space regardless of the size of the input.
- The only additional space used is for a constant number of pointers (`prev`, `current`, `next_node`).

### Recursive Approach:

**Time Complexity:**
- The time complexity is $O(n)$, where n is the number of nodes in the linked list.
- In each recursive call, only one node is processed, and there are $n$ recursive calls.

**Space Complexity:**
- The space complexity is $O(n)$ due to the function call stack.
- In the worst case, the function call stack can grow to be as large as the number of nodes in the linked list.

### Summary:

- Both approaches have the same time complexity, but the iterative approach has a more favorable constant factor and is often more efficient in practice.
- The iterative approach has a better space complexity of $O(1)$, making it a preferred choice for reversing linked lists in terms of space efficiency.
- The recursive approach, while elegant, may lead to stack overflow for very long linked lists and is less memory-efficient.
