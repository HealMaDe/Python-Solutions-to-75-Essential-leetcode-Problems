## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a two-pointer approach to find the middle node of the linked list.

1. use two pointers, pointing to the head of the linked list.

2. Move at Different Speeds

3. Delete the Middle Node

4. checks for edge cases where the linked list is empty or contains only one node



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def deleteMiddle(self, head):
        if not head or not head.next:
            return None  # No middle node to delete

        slow = head
        fast = head
        prev = None

        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next

        # Delete the middle node
        if prev:
            prev.next = slow.next
        else:
            head = head.next  # If the head is the only node

        return head
```

```python
#Example 1
head = ListNode(1, ListNode(3, ListNode(4, ListNode(7, ListNode(1, ListNode(2, ListNode(6)))))))
result = Solution().deleteMiddle(head)
print(result)
```

```python
#Example 2
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
result = Solution().deleteMiddle(head)
print(result)
```

```python
#Example 3
head = ListNode(2, ListNode(1))
result = Solution().deleteMiddle(head)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

This Solution has a time complexity of $O(n)$, where $n$ is the number of nodes in the linked list, as it requires traversing the entire list once.

The space complexity is $O(1)$, as it uses only a constant amount of extra space regardless of the size of the input.
