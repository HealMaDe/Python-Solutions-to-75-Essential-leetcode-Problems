## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a simple approach and straightforward algorithm to reverse the words in a string by splitting it into individual words, reversing their order, and then joining them back together.. Here's a breakdown of the Algorithm:

1. Split the string s into a list of words using the split() method. By default, split() splits the string at whitespace characters.

2. Reverse the order of the words in the list using the reverse() method.

3. Join the reversed list of words back into a string using the join() method, with a space as the separator.

4. Return the resulting reversed string.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def reverseWords(self, s):
        L = s.split()
        L.reverse()
        return " ".join(L)
```

```python
#Example 1
result = Solution().reverseWords("the sky is blue")
print(result)
```

```python
#Example 2
result = Solution().reverseWords("  hello world  ")
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `reverseWords` function can be analyzed as follows:

1. Time Complexity:
   - The `split()` method takes $O(n)$ time, where $n$ is the length of the input string `s`. It splits the string into words based on whitespace.
   - The `reverse()` method operates on the list of words and takes $O(n)$ time as well, where $n$ is the number of words in the list.
   - The `join()` method takes $O(n)$ time, where $n$ is the total length of the words in the list.
   - Therefore, the overall time complexity is $O(n)$, where $n$ is the length of the input string `s`.

2. Space Complexity:
   - The space complexity is determined by the variables used and the additional space required for the list of words.
   - The `split()` method creates a list of words, which requires $O(n)$ space to store the individual words.
   - The `reverse()` method operates on the existing list in-place and does not consume additional space.
   - The `join()` method creates a new string with the reversed words, which requires $O(n)$ space.
   - Therefore, the overall space complexity is $O(n)$, where $n$ is the length of the input string `s`.
