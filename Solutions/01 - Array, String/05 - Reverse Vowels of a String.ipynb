## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

To solve the problem of reversing only the vowels in a given string, I utilize the following algorithm:

1. Initialize two pointers, start and end, pointing to the beginning and end of the string, respectively.

2. Convert the string to a list of characters to make the necessary modifications.

3. While start is less than or equal to end, perform the following steps:

  - Check if the character at start is a vowel. If it is not a vowel, increment start by 1.
  - Check if the character at end is a vowel. If it is not a vowel, decrement end by 1.
  - If both characters at start and end are vowels, swap them.
  - Increment start by 1 and decrement end by 1.
8. Convert the modified list of characters back into a string and return it.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def reverseVowels(self, s):
        s_list = list(s)
        start, end = 0, len(s_list) - 1

        while start <= end:
            if s_list[start] not in 'aeiouAEIOU':
                start += 1
            elif s_list[end] not in 'aeiouAEIOU':
                end -= 1
            else:
                s_list[start], s_list[end] = s_list[end], s_list[start]
                start += 1
                end -= 1

        return ''.join(s_list)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `reverseVowels` function can be analyzed as follows:

1. Time Complexity:
   - The while loop iterates until `start` becomes greater than `end`. In the worst case, it iterates through half of the string.
   - Inside the loop, the checks `s_list[start] not in 'aeiouAEIOU'` and `s_list[end] not in 'aeiouAEIOU'` both take constant time.
   - Swapping characters `s_list[start]` and `s_list[end]` also takes constant time.
   - Therefore, the overall time complexity of your function is $O(n/2)$, which simplifies to $O(n)$, where $n$ is the length of the input string `s`.

2. Space Complexity:
   - The space complexity is determined by the variables used and the list `s_list`.
   - The variables `start` and `end` require constant space.
   - The conversion of the string `s` to a list `s_list` takes $O(n)$ space, where $n$ is the length of `s`.
   - The `join` operation in the return statement also takes $O(n)$ space to create the final string.
   - Therefore, the overall space complexity is $O(n)$.
