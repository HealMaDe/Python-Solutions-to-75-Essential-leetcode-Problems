## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a single-pass approach.

Here's the algorithm:

1. Initialize three variables: first with positive infinity, second with positive infinity, and third with iteration through nums.
2. Iterate through each element number in the nums array.
  - If num is less than first, update first with num.
  - If num is greater than first and less than second, update second with num.
  - If num is greater than second, we have found a triplet, so return True.
3. If no triplet is found, return False.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def increasingTriplet(self, nums):
        min1 = min2 = float("inf")
        for number in nums:
            if number <= min1:
                min1 = number
            elif number <= min2:
                min2 = number
            else:
                return True
        else:
            return False
```

```python
#Example 1
nums = [1,2,3,4,5]
result = Solution().increasingTriplet(nums)
print(result)
```

```python
#Example 2
nums = [5,4,3,2,1]
result = Solution().increasingTriplet(nums)
print(result)
```

```python
#Example 3
nums = [2,1,5,0,4,6]
result = Solution().increasingTriplet(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(n)$, where $n$ is the length of the input array `nums`. This is because the code iterates through each element of the array once, performing constant time operations for each element.

The space complexity is $O(1)$ as it uses a constant amount of extra space. The variables `min1` and `min2` are the only additional variables used, and they take constant space regardless of the size of the input array.
