## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

The approach I use is straightforward. It involves iterating over the two input strings simultaneously and appending characters alternately to the resulting merged string.

Here's a breakdown of the Algorithm:

1. Initialize an empty string to store the merged result.
2. Initialize two pointers to keep track of the current positions in `word1` and `word2`, respectively.
3. Use a while loop to iterate as long as both pointers are within the bounds of their respective strings.
4. Inside the loop, concatenate the character to the merged string.
5. Increment both pointers by 1 to move to the next character in each string.
6. After the loop, append the remaining characters from `word1` and `word2` to the merged string, using slicing.
7. Return the final merged string.

This approach ensures that characters from both input strings are merged alternately until one of the strings is exhausted. The remaining characters from the longer string are then appended to the end of the merged string.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        merged = ""
        i = 0
        j = 0

        while i < len(word1) and j < len(word2):
            merged += word1[i] + word2[j]
            i += 1
            j += 1

        merged += word1[i:] + word2[j:]

        return merged
```

```python
#Example 1
word1 ,  word2= "abc", "pqr"
result = Solution().mergeAlternately(word1, word2)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `mergeAlternately` function can be analyzed as follows:

1. Time Complexity:
   - The while loop iterates until either `i` reaches the end of `word1` or `j` reaches the end of `word2`. Therefore, the number of iterations is determined by the length of the shorter string between `word1` and `word2`.
   - Thus, the time complexity of the while loop is $O(min(len(word1), len(word2)))$.
   - The concatenation of strings `merged += word1[i] + word2[j]` inside the loop takes constant time.
   - The concatenation of the remaining characters `word1[i:]` and `word2[j:]` after the loop also takes $O(len(word1) - i + len(word2) - j)$ time.
   - Combining these, the overall time complexity of your function is $O(min(len(word1), len(word2))) + O(len(word1) - i + len(word2) - j)$.
   - Since the second term is dominated by the first term, the final time complexity is $O(min(len(word1), len(word2)))$ or simply $O(n)$.

2. Space Complexity:
   - The space complexity of your function is determined by the `merged` string that holds the merged result.
   - The length of the `merged` string will be equal to the sum of the lengths of `word1` and `word2`.
   - Therefore, the space complexity of your function is $O(len(word1) + len(word2))$.
