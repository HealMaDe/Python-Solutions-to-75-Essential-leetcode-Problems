## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

To solve this problem and generate the array answer in O(n) time complexity without using division, I utilize the concept of prefix and suffix products. I also solve the problem with O(1) extra space complexity by utilizing the output array itself to store the intermediate results.

Here's the algorithm:
1. Initialize the answer array with the same length as nums and fill it with 1's.
2. Calculate the prefix products and store them in the answer array. Starting from index 1.
3. Initialize a variable suffix product to 1.
4. Iterate through the nums array from right to left.
  - Calculate the suffix product and update answer.
  - Update suffix product by multiplying it with nums.
5. Return the answer array.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def productExceptSelf(self, nums):
        n = len(nums)
        answer = [1] * n

        # Calculate prefix products
        for i in range(1, n):
            answer[i] = answer[i-1] * nums[i-1]

        # Calculate suffix products and update answer
        suffix_product = 1
        for i in range(n-1, -1, -1):
            answer[i] *= suffix_product
            suffix_product *= nums[i]

        return answer
```

```python
#Example 1
nums = [1,2,3,4]
result = Solution().productExceptSelf(nums)
print(result)
```

```python
#Example 2
nums = [-1,1,0,-3,3]
result = Solution().productExceptSelf(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of `productExceptSelf` is $O(n)$, where $n$ is the length of the input array `nums`. This is because there are two separate loops that iterate through the array: one for calculating the prefix products and another for calculating the suffix products.

The space complexity of the code is $O(1)$.  The space used by the algorithm is only for the output array itself. Hence, it has $O(1)$ extra space complexity.
