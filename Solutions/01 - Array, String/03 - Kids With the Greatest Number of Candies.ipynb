## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

The approach I use is a straightforward and efficient algorithm. It involves finding the maximum number of candies in the given list and comparing each kid's number of candies with the maximum value.

Here's a breakdown of the Algorithm:

1. Initialize a list result of the same length as the candies list, filled with False values. This list will store whether each kid can have the greatest number of candies.

2. Iterate over the indices of the candies list using a for loop.

3. For each index, compare the sum of candies for each kid and extraCandies with the maximum value in the candies list.

4. If the value is greater than or equal to the maximum value, set the index result to True, indicating that the kid can have the greatest number of candies.

5. After iterating through all the indices, return the result list.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        result = [False]*len(candies)

        for i in range(len(candies)):
            if candies[i] + extraCandies >= max(candies):
                result[i] = True

        return result
```

```python
#Example 1
candies = [2,3,5,1,3]
result = Solution().kidsWithCandies(candies, extraCandies = 3)
print(result)
```

```python
#Example 2
candies = [4,2,1,1,2]
result = Solution().kidsWithCandies(candies, extraCandies = 1)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `kidsWithCandies` function can be analyzed as follows:

1. Time Complexity:
   - The line `result = [False]*len(candies)` takes $O(len(candies))$ time to create a list of the same length as the `candies` list.
   - The for loop iterates over the range of `len(candies)`, which takes $O(len(candies))$ time.
   - Inside the loop, the comparison `candies[i] + extraCandies >= max(candies)` takes $O(len(candies))$ time since `max(candies)` iterates over the `candies` list to find the maximum value.
   - Therefore, the overall time complexity is $O(len(candies)) + O(len(candies)) * O(len(candies)) = O(len(candies)^2)$.

2. Space Complexity:
   - The space complexity is determined by the `result` list, which has a length equal to the `candies` list. Thus, it requires $O(len(candies))$ space.
