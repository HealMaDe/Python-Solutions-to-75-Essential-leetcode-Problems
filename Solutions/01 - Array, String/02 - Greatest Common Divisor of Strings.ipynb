## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

The approach I used is based on the observation that if two strings, `str1` and `str2`, have a common divisor, they can be concatenated in any order to form a common multiple.

Here's a breakdown of the Algorithm:

1. You check if `str1 + str2` is equal to `str2 + str1`. This condition verifies if `str1` and `str2` have a common divisor. If they don't, it means there is no solution, and you return an empty string.

2. If the strings pass the common divisor check, you proceed to calculate the greatest common divisor (GCD) of the lengths of `str1` and `str2`. This step is performed using the Euclidean algorithm, where you repeatedly update lengths of strings until length of `str2` becomes zero. The GCD is then stored.

3. Finally, you return a substring of `str1` from index 0 to gcd. This substring represents the greatest common divisor string of `str1` and `str2`.

**Note:**

The Euclidean algorithm is a way to find the greatest common divisor of two positive integers, $a$ and $b$.

Formal description of the algorithm:
- Input Two positive integers, $a$ and $b$.
- Output The greatest common divisor, $g$, of $a$ and $b$.
- Internal computation
  - If $a < b$, exchange $a$ and $b$.
  - Divide $a$ by $b$ and get the remainder, $r$. If $r = 0$, report $b$ as the GCD of $a$ and $b$.
  - Replace $a$ by $b$ and replace $b$ by $r$. Return to the previous step.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution():
    def gcdOfStrings(self, str1, str2):
        if str1 + str2 != str2 + str1:
            return ""

        len1 = len(str1)
        len2 = len(str2)

        #Euclidean algorithm
        while len2!=0:
            len1 , len2 = len2 , len1 % len2
        gcd = len1

        return str1[:gcd]
```

```python
#Example 1
str1 = "ABCABC"
str2 = "ABC"
result = Solution().gcdOfStrings(str1, str2)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `gcdOfStrings` function can be analyzed as follows:

1. Time Complexity:
   - The check `str1 + str2 != str2 + str1` takes $O(len(str1) + len(str2))$ time since it concatenates the strings and compares them.
   - The Euclidean algorithm loop iterates until `len2` becomes zero, which takes $O(log(min(len(str1), len(str2))))$ iterations.
   - Inside the loop, the calculation `len1, len2 = len2, len1 % len2` takes constant time.
   - Therefore, the overall time complexity is $O(len(str1) + len(str2)) + O(log(min(len(str1), len(str2))))$.

2. Space Complexity:
   - The space complexity of function is determined by the variables used.
   - The variables `len1` and `len2` require constant space.
   - The variable `gcd` also requires constant space.
   - Therefore, the space complexity is $O(1)$.
