## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

The approach I use is a greedy algorithm that iterates over the flowerbed to determine if it's possible to place `n` flowers in it.

Here's a breakdown of the Algorithm:

1. Initialize a variable to keep track of the number of flowers placed. Set it to 0 initially.

2. Initialize a pointer to 0, representing the current index in the flowerbed.

3. Use a while loop that continues until reaches the end of the flowerbed.

4. Inside the loop, check if the current location is empty (0).

5. If the current location is empty, check the following conditions to determine if a flower can be placed:
   - Check if pointer is at the beginning of the flowerbed or if the previous location is also empty (0).
   - Check if pointer is at the end of the flowerbed or if the next location is also empty (0).

6. If both conditions are satisfied, it means a flower can be placed at the current location. Set flowerbed to 1 to indicate a flower has been placed.

7. Increment pointer to move to the next location in the flowerbed.

8. After the loop ends, check if the number of flowers placed is greater than or equal to the required number of flowers `n`. Return `True` if it's true and `False` otherwise.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        count = 0
        i = 0

        while i < len(flowerbed):
            if flowerbed[i] == 0:
                if (i == 0 or flowerbed[i - 1] == 0):
                    if (i == len(flowerbed) - 1 or flowerbed[i + 1] == 0):
                        flowerbed[i] = 1
                        count += 1
            i += 1

        return count >= n
```

```python
#Example 1
flowerbed = [1,0,0,0,1]
result = Solution().canPlaceFlowers(flowerbed, n=1)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of `canPlaceFlowers` function can be analyzed as follows:

1. Time Complexity:
   - The while loop iterates until `i` reaches the length of the `flowerbed`, which takes $O(len(flowerbed))$ iterations.
   - Inside the loop, the checks `flowerbed[i] == 0`, `(i == 0 or flowerbed[i - 1] == 0)`, and `(i == len(flowerbed) - 1 or flowerbed[i + 1] == 0)` all take constant time.
   - Assigning `flowerbed[i] = 1` and incrementing `count` both take constant time as well.
   - Therefore, the overall time complexity is $O(len(flowerbed))$.

2. Space Complexity:
   - The space complexity is determined by the variables used.
   - The variables `count` and `i` require constant space.
   - Therefore, the space complexity is $O(1)$.
