## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a two-pointer approach to compress the characters in-place.

Here's the step-by-step algorithm:

1. Initialize two pointers: write and read to 0.
2. Initialize a variable count to 1, representing the count of consecutive repeating characters.
3. Iterate through the characters in the chars array, starting from the second character.
4. If the current character is equal to the previous character, increment the count variable.
5. If the current character is different from the previous character:
  - Write the previous character to the chars array.
  - Increment the index by 1.
  - If the count is greater than 1, convert the count to a string and write each digit to the chars array at the corresponding indices.
  - Reset the count to 1.
6. After the loop ends, write the last character and its count (if greater than 1) to the chars array.
7. Return the index, which represents the length of the compressed array.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def compress(self, chars):
        answer = 0  # Initialize the variable to keep track of the write index
        count = 1   # Initialize the variable to count consecutive repeating characters

        for i in range(1, len(chars)):
            if chars[i] == chars[i - 1]:  # If the current character is the same as the previous one
                count += 1  # Increment the count of consecutive repeating characters
            else:
                chars[answer] = chars[i - 1]  # Write the previous character to the answer position
                answer += 1  # Increment the write index

                if count > 1:  # If there are more than one consecutive repeating characters
                    count_str = str(count)
                    for digit in count_str:  # Write each digit of the count as a separate character
                        chars[answer] = digit
                        answer += 1

                count = 1  # Reset the count for the new group of consecutive characters

        chars[answer] = chars[-1]  # Write the last character to the answer position
        answer += 1  # Increment the write index

        if count > 1:  # If there are more than one consecutive repeating characters
            count_str = str(count)
            for digit in count_str:  # Write each digit of the count as a separate character
                chars[answer] = digit
                answer += 1

        return answer  # Return the length of the compressed array
```

```python
#Example 1
chars = ["a","a","b","b","c","c","c"]
result = Solution().compress(chars)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this algorithm is $O(n)$, where $n$ is the length of the chars array. This is because I perform a single pass through the array. The space complexity is $O(1)$ since we are modifying the chars array in-place without using any additional data structures.
