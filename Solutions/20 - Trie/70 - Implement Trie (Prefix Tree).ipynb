## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I create a Trie instance and insert the word "apple" into it. Then I perform search operations on the Trie using the search method and check for prefixes using the startsWith method. The results match the expected output specified in the test case.

The implementation consists of two classes: TrieNode and Trie.

1.  The TrieNode class represents a node in the Trie. It has a dictionary children that maps characters to their corresponding child nodes, and a boolean is_word flag to indicate if the node represents the end of a word.

2. The Trie class represents the Trie itself.
  - It has a root node that represents an empty string.
  - The insert method is used to insert a word into the Trie by traversing the characters of the word and creating new nodes if necessary.
  - The search method is used to search for a complete word in the Trie, returning True if the word exists and False otherwise.
  - The startsWith method is used to check if a prefix exists in the Trie, returning True if the prefix is found and False otherwise.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class TrieNode:
    def __init__(self):
        self.children = {}  # Dictionary to store character-child node mappings
        self.is_word = False  # Flag to indicate if the node represents the end of a word


class Trie:
    def __init__(self):
        self.root = TrieNode()  # Create the root node of the trie

    def insert(self, word):
        node = self.root  # Start traversal from the root node
        for char in word:
            if char not in node.children:  # If the character is not in the children, create a new node
                node.children[char] = TrieNode()
            node = node.children[char]  # Move to the child node
        node.is_word = True  # Mark the last node as the end of a word

    def search(self, word):
        node = self.root  # Start traversal from the root node
        for char in word:
            if char not in node.children:  # If the character is not in the children, the word does not exist
                return False
            node = node.children[char]  # Move to the child node
        return node.is_word  # Return True if the last node represents the end of a word

    def startsWith(self, prefix):
        node = self.root  # Start traversal from the root node
        for char in prefix:
            if char not in node.children:  # If the character is not in the children, the prefix does not exist
                return False
            node = node.children[char]  # Move to the child node
        return True  # Return True if the traversal completes successfully without encountering missing characters
```

```python
trie = Trie()
trie.insert("apple")
print(trie.search("apple"))   # Output: True
print(trie.search("app"))     # Output: False
print(trie.startsWith("app")) # Output: True
trie.insert("app")
print(trie.search("app"))     # Output: True
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:
- Insertion: The time complexity for inserting a word into the Trie is $O(k)$, where $k$ is the length of the word. This is because we need to traverse each character of the word and potentially create new nodes if they don't exist already. Therefore, the time complexity for insertion is proportional to the length of the word being inserted.
- Search: The time complexity for searching a word in the Trie is also $O(k)$, where $k$ is the length of the word. We traverse each character of the word, following the corresponding child nodes. If at any point, a character is not found in the children of the current node, the search terminates. Hence, the time complexity is proportional to the length of the searched word.
- Starts With: The time complexity for checking if a prefix exists in the Trie is $O(p)$, where $p$ is the length of the prefix. We traverse each character of the prefix, following the corresponding child nodes. If at any point, a character is not found in the children of the current node, the prefix does not exist. Thus, the time complexity is proportional to the length of the prefix.

Space Complexity:

The space complexity of the Trie data structure depends on the number of unique characters in the dataset and the number of words inserted into the Trie.
- The space required for storing the Trie nodes is proportional to the number of characters in the dataset. In the worst case, if all unique characters are present in the dataset, the space complexity is $O(26)$ or $O(52)$ for lowercase and uppercase English letters, respectively.
- The space required for storing the inserted words depends on the length of the words and their prefixes. If there are many words with common prefixes, the Trie can be more space-efficient compared to other data structures. However, in the worst case, if all inserted words have distinct prefixes, the space complexity can be proportional to the total length of all the words.
