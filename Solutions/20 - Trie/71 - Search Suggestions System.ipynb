## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a combination of sorting and binary search to efficiently find the suggestions for each prefix being typed. By leveraging the lexicographically sorted products list, we can quickly locate the range of products with a common prefix, ensuring that the resulting suggestions are accurate and limited to three products at most.

Here are the steps of the algorithm:

1. Sort the products list lexicographically. This step ensures that the products are in the correct order for binary search.

2. Create an empty list to store the suggestions for each character of the search word.

3.  Initialize an empty string to keep track of the current prefix being typed.

4. Iterate over each character in the search word.

5. Concatenate the current character with the prefix to build the current prefix being typed.

6. Use binary search to find the leftmost index of the prefix in the sorted products list.

7. Use binary search to find the rightmost index of the prefix concatenated with a special character (e.g., '~') in the sorted products list.

8. Retrieve the products with the common prefix by slicing the products list from the left index to the right index.

9. Take the first three products from the retrieved list to limit the suggestions to a maximum of three.

10. Append the matched products to the suggestions list.

11. Repeat steps 5 to 10 for each character of the search word.

12. Return the suggestions list as the final result.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
import bisect
class Solution(object):
    def suggestedProducts(self, products, searchWord):
        products.sort()
        suggestions = []
        prefix = ""
        for char in searchWord:
            prefix += char
            left = bisect.bisect_left(products, prefix)
            right = bisect.bisect_right(products, prefix + "~")
            matched_products = products[left:right][:3]
            suggestions.append(matched_products)
        return suggestions
```

```python
#Example 1
products = ["mobile","mouse","moneypot","monitor","mousepad"]
searchWord = "mouse"
result = Solution().suggestedProducts(products, searchWord)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Sorting the products list: $O(n *log (n))$, where $n$ is the length of the products list.
- For each character in the search word, we perform two binary search operations using `bisect_left` and `bisect_right`, each taking $O(log (n))$ time in the worst case.
- Retrieving at most three products from the sliced range takes constant time.
- Overall, the time complexity is $O(n *log (n) + m *log (n))$, where $m$ is the length of the search word.

Space Complexity:
- The space complexity is $O(n)$ to store the sorted products list.
- The additional space required for storing the suggestions and the prefix is $O(m)$, where $m$ is the length of the search word.
- Overall, the space complexity is $O(n + m)$.
