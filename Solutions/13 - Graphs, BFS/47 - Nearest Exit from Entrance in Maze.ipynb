## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a Breadth-First Search (BFS) algorithm. We'll start from the entrance cell and explore neighboring cells until we find an exit cell or exhaust all possibilities.

Here it goes:

1. Initialize a queue to store the cells to visit and a set to keep track of visited cells.
2. Enqueue the entrance cell into the queue and mark it as visited.
3. Initialize a variable `steps` to 0 to keep track of the number of steps taken.
4. While the queue is not empty:
   - Increment `steps` by 1.
   - Get the size of the queue (this will be the number of cells to process at the current level).
   - Iterate through each cell in the current level:
     - Dequeue a cell from the front of the queue.
     - Check if the cell is an exit cell (at the border of the maze). If yes, return `steps` as the shortest path length.
     - Explore all unvisited neighboring cells (up, down, left, right) that are not walls ('+') and enqueue them into the queue. Mark them as visited.
5. If no exit cell is found after the BFS, return -1 to indicate that no path exists.





## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from collections import deque
class Solution(object):
    def nearestExit(self, maze, entrance):
        m = len(maze)
        n = len(maze[0])
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]  # Possible directions: right, left, down, up

        queue = deque()
        queue.append(entrance)
        visited = set()
        visited.add(tuple(entrance))
        steps = 0

        while queue:
            steps += 1
            level_size = len(queue)

            for _ in range(level_size):
                row, col = queue.popleft()

                for dx, dy in directions:
                    new_row = row + dx
                    new_col = col + dy

                    if 0 <= new_row < m and 0 <= new_col < n and maze[new_row][new_col] == '.' and (new_row, new_col) not in visited:
                        if new_row == 0 or new_row == m - 1 or new_col == 0 or new_col == n - 1:
                            return steps

                        queue.append((new_row, new_col))
                        visited.add((new_row, new_col))

        return -1
```

```python
# Example 1
maze = [["+","+",".","+"],[".",".",".","+"],["+","+","+","."]]
entrance = [1,2]
result = Solution().nearestExit(maze,entrance)
print(result)
```

```python
# Example 2
maze = [["+","+","+"],[".",".","."],["+","+","+"]]
entrance = [1,0]
result = Solution().nearestExit(maze,entrance)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
The BFS algorithm explores each cell in the maze once. In the worst case, every cell can be visited, resulting in a time complexity of $O(m * n)$, where $m$ and $n$ are the dimensions of the maze.

Space Complexity:
The space complexity is determined by the space used to store the queue, visited set, and other variables. In the worst case, when all cells are visited, the maximum number of elements in the queue would be the total number of cells in the maze. Therefore, the space complexity is $O(m * n)$ to store the queue.

Additionally, the visited set stores the visited cells, which also contributes to the space complexity. In the worst case, all cells can be visited, resulting in a space complexity of $O(m * n)$ for the visited set.

Hence, the overall space complexity is $O(m * n)$.
