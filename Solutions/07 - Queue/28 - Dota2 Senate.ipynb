## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a greedy strategy. I simulate the process of banning senators' rights until one party has all its senators' rights banned, and the other party wins.

1. use two queues to keep track of the indices of senators from each party.
2. In each round, iterate through the senators
  - decide whether to ban the senator's right or not based on the order of banning.
  - If a senator's right is banned, skip the senator.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from collections import deque
class Solution(object):
    def predictPartyVictory(self, senate):
        radiant_queue = deque()
        dire_queue = deque()
        n = len(senate)

        # Initialize the queues based on the party of each senator
        for i, party in enumerate(senate):
            if party == 'R':
                radiant_queue.append(i)
            else:
                dire_queue.append(i)

        # Simulate the banning process
        while radiant_queue and dire_queue:
            radiant_index = radiant_queue.popleft()
            dire_index = dire_queue.popleft()

            # Decide which senator's right to ban
            if radiant_index < dire_index:
                radiant_queue.append(radiant_index + n)
            else:
                dire_queue.append(dire_index + n)

        # Determine the winner based on the remaining senators
        return "Radiant" if radiant_queue else "Dire"
```

```python
#Example 1
senate = "RD"
result = Solution().predictPartyVictory(senate)
print(result)
```

```python
#Example 2
senate = "RDD"
result = Solution().predictPartyVictory(senate)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Let's analyze the time and space complexity of the my solution:

### Time Complexity:
The time complexity is $O(n)$, where $n$ is the length of the input string `senate`. In each round, we either ban a senator's right ($O(1)$ operation) or skip a senator whose right has been banned. The process continues until one party wins. The number of rounds is at most N, as each senator can be banned at most once.

### Space Complexity:
The space complexity is $O(n)$, where $n$ is the length of the input string `senate`. The space is used to store the indices of senators in the queues. In the worst case, all senators are enqueued in one of the queues, leading to a linear amount of space.
