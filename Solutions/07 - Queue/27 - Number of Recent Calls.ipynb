## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

1. Use a queue to keep track of the recent requests within the specified time frame
2. When a new request comes in, enqueue its timestamp
  - remove any outdated requests from the front of the queue
3. ensure the timestamps within the last 3000 milliseconds are retained.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from collections import deque
class RecentCounter(object):
    def __init__(self):
        self.requests = deque()

    def ping(self, t):
        # Enqueue the current timestamp
        self.requests.append(t)

        # Remove outdated requests from the front of the queue
        while self.requests and self.requests[0] < t - 3000:
            self.requests.popleft()

        # Return the number of requests within the last 3000 milliseconds
        return len(self.requests)
```

```python
# Example
recentCounter = RecentCounter()
print(recentCounter.ping(1))    # Output: 1
print(recentCounter.ping(100))  # Output: 2
print(recentCounter.ping(3001)) # Output: 3
print(recentCounter.ping(3002)) # Output: 3
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


### Time Complexity:
1. **Initialization (`__init__` method):** The initialization involves creating an empty deque, which is a constant-time operation. Therefore, the time complexity of initialization is $O(1)$.

2. **Ping (`ping` method):**
   - Enqueuing the current timestamp is a constant-time operation.
   - Removing outdated requests involves dequeuing elements from the front of the deque. In the worst case, we may need to dequeue all elements until we find the first valid request. However, each request is enqueued and dequeued at most once. Therefore, the amortized time complexity of dequeuing is $O(1)$.
   
   As a result, the overall time complexity of the `ping` method is $O(1)$.

### Space Complexity:
The space complexity is $O(n)$, where $n$ is the number of requests stored in the deque. In the worst case, the deque stores all requests within the last 3000 milliseconds. The space complexity is directly proportional to the number of requests within the time frame.
