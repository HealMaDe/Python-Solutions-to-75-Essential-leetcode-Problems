## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I utilize a modified BFS approach.

1. initialize a queue and enqueue the root node
2. initialize variables to keep track of the current level, the maximum level with the maximal sum, and the maximum sum found
3. perform a level-by-level traversal using the queue.
4. At each level, calculate the sum of node values and compare it with the current maximum sum.
5. If the sum is greater, update the maximum sum and the corresponding level

6. After processing all levels, return the level with the maximum sum

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from collections import deque

# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root:
            return 0

        queue = deque()
        queue.append(root)
        level = 1
        max_level = 1
        max_sum = float('-inf')

        while queue:
            level_sum = sum(node.val for node in queue)
            if level_sum > max_sum:
                max_sum = level_sum
                max_level = level

            level += 1
            size = len(queue)

            for _ in range(size):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)

        return max_level
```

```python
# Example 1
# Create a sample tree
root = TreeNode(1)
root.left = TreeNode(7)
root.right = TreeNode(0)
root.left.left = TreeNode(7)
root.left.right = TreeNode(-8)
result = Solution().maxLevelSum(root)
print(result)
```

```python
#Example 2
# Create a sample tree
root = TreeNode(989)
root.right = TreeNode(10250)
root.right.left = TreeNode(98693)
root.right.right = TreeNode(-89388)
root.right.right.right = TreeNode(-32127)
result = Solution().maxLevelSum(root)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
The solution performs a level-by-level traversal of the binary tree using a modified BFS approach. In the worst case, where all nodes are visited, the time complexity is $O(N)$, where $N$ is the number of nodes in the tree.

Space Complexity:
The space complexity is determined by the queue used for the BFS traversal. In the worst case, the maximum number of nodes in the queue would be at the maximum level of the binary tree. This can occur in a completely unbalanced binary tree, where the number of nodes at the maximum level would be roughly $N/2$. Therefore, the space complexity is $O(N)$.

Note: If the binary tree is balanced, the number of nodes at the maximum level would be much smaller, resulting in a better space complexity of $O(log (N))$ in terms of the height of the binary tree.
