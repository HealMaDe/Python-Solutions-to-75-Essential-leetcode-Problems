## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>


I use two main attributes: heap and min_num. The heap is used to store the elements in ascending order, and min_num keeps track of the smallest number that has not been added to the heap.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
import heapq

class SmallestInfiniteSet:
    def __init__(self):
        self.heap = []
        self.min_num = 1

    def popSmallest(self):
        if self.heap:
            return heapq.heappop(self.heap)
        self.min_num += 1
        return self.min_num - 1

    def addBack(self, num):
        if self.min_num > num and num not in self.heap:
            heapq.heappush(self.heap, num)
```

```python
# Example 1
obj = SmallestInfiniteSet()
obj.addBack(2)
print(obj.popSmallest())  # Output: 1
print(obj.popSmallest())  # Output: 2
print(obj.popSmallest())  # Output: 3
obj.addBack(1)
print(obj.popSmallest())  # Output: 1
print(obj.popSmallest())  # Output: 4
print(obj.popSmallest())  # Output: 5
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- The `popSmallest` method has a time complexity of $O(1)$ when the heap is not empty since it performs a constant-time operation to pop the smallest element from the heap.
- If the heap is empty, the `popSmallest` method increments `min_num` and returns the previous `min_num - 1`, which is also a constant-time operation.
- The `addBack` method has a time complexity of $O(log (n))$ in the worst case, where $n$ is the number of elements in the heap. This is because it uses `heapq.heappush`, which performs a logarithmic-time operation to maintain the heap property.

Overall, the time complexity of the code depends on the number of operations performed. If there are m `addBack` operations and n `popSmallest` operations, the overall time complexity would be $O(m *log (n) + n)$, considering the worst-case scenario of `heapq.heappush` and `heapq.heappop` operations.

Space Complexity:
- The space complexity of the code is $O(n)$, where $n$ is the number of elements in the heap. This is because the `heap` attribute stores the elements, and its space usage grows linearly with the number of elements.
- Additionally, the `min_num` attribute occupies constant space.
