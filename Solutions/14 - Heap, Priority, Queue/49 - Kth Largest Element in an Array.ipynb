## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I solve the problem without sorting the entire array by utilizing the concept of a min heap.

Here's an algorithm to find the kth largest element in an array `nums`:

1. Create a min heap with the first k elements of the array `nums`.
2. For the remaining elements in `nums` (starting from index k):
   - If the current element is greater than the root of the min heap:
     - Remove the root element from the heap.
     - Insert the current element into the heap.
3. After processing all the elements in `nums`, the root of the min heap will contain the kth largest element.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
import heapq
class Solution(object):
    def findKthLargest(self, nums, k):
        heap = []

        # Create a min heap with the first k elements
        for i in range(k):
            heapq.heappush(heap, nums[i])

        # Process the remaining elements
        for i in range(k, len(nums)):
            if nums[i] > heap[0]:
                heapq.heappop(heap)
                heapq.heappush(heap, nums[i])

        # The root of the heap will be the kth largest element
        return heap[0]
```

```python
# Example 1
nums = [3,2,1,5,6,4]
k = 2
result = Solution().findKthLargest(nums,k)
print(result)
```

```python
# Example 2
nums = [3,2,3,1,2,4,5,5,6]
k = 4
result = Solution().findKthLargest(nums,k)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time complexity:

1. Creating the initial min heap: $O(k * log (k))$ - We insert the first k elements into the min heap, which takes $O(log k)$ time for each insertion. Since we perform $k$ insertions, the total time complexity is $O(k * log (k))$.

2. Processing the remaining elements: $O((n - k) * log (k))$ - We iterate over the remaining elements in `nums` (from index $k$ to $n-1$) and compare each element with the root of the min heap. If the element is larger, we perform a heap pop and push operation, each taking $O(log (k))$ time. Since there are $(n - k)$ remaining elements, the total time complexity is $O((n - k) * log (k))$.

Hence, the overall time complexity is $O(k * log (k) + (n - k) * log (k)) = O(n * log (k))$.

The space complexity of the algorithm is $O(k)$, as we only store the k elements in the min heap.
