## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>


I use a greedy approach to solve the problem. I create a list `L` that combines the elements of `nums1` and `nums2` into pairs. The list `L` is then sorted in descending order based on the second element of each pair (i.e., `j`).

The code uses a heap to maintain a sliding window of size `k` over the elements of `nums1`. It iterates over the sorted list `L` and performs the following steps:

1. Pushes the first element of the pair (i.e., `i`) onto the heap.
2. Updates the `total` by adding the current element (`i`) to it.
3. If the size of the heap is equal to `k`, calculates the current score (`total * j`) and updates the `best` score if the current score is greater than `best`.
4. If the size of the heap exceeds `k`, removes the smallest element from the heap and subtracts it from the `total`.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from heapq import heappush, heappop
class Solution(object):
    def maxScore(self, nums1, nums2, k):
        L = []

        for i , j in zip(nums1,nums2):
            L.append([i,j])

        L.sort(key=lambda x:-x[1])
        heap=[]
        best = total = 0

        for i , j in L:
            heappush(heap,i)
            total += i
            if len(heap) == k:
                best = max(best,total*j)
                total -= heappop(heap)

        return best
```

```python
# Example 1
nums1 = [1,3,3,2]
nums2 = [2,1,3,4]
k = 3
result = Solution().maxScore(nums1, nums2, k)
print(result)
```

```python
# Example 2
nums1 = [4,2,3,1,1]
nums2 = [7,5,10,9,6]
k = 1
result = Solution().maxScore(nums1, nums2, k)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Constructing the list `L` takes $O(n)$ time, where $n$ is the length of `nums1` (assuming both `nums1` and `nums2` have the same length).
- Sorting the list `L` takes $O(n *log (n))$ time.
- The heap operations (pushing and popping elements) are performed in $O(log (k))$ time, where $k$ is the given value.
- Overall, the time complexity of the code is $O(n *log (n) + k *log (k))$.

Space Complexity:
- The space complexity of the code is $O(n)$ because the list `L` stores pairs of elements from `nums1` and `nums2`.
- The heap stores at most `k` elements, resulting in an additional space complexity of $O(k)$.
- Overall, the space complexity is $O(n + k)$.
