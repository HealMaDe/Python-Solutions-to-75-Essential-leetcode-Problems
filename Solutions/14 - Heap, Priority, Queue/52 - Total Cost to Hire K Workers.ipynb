## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>


1. Initialize two heaps: `min_heap_first_candidates` for the first `candidates` workers and `max_heap_last_candidates` for the last `candidates` workers.
2. Use a loop to perform k hiring sessions.
3. In each session, choose the worker with the lowest cost from either `min_heap_first_candidates` or `max_heap_last_candidates`.
4. Adjust indices (`i` and `j`) and update the heaps accordingly.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
from heapq import heapify, heappop, heappush

class Solution:
    def totalCost(self, costs, k, candidates):
        # Separate the first 'candidates' and last 'candidates' costs into two heaps
        min_heap_first_candidates = costs[:candidates]
        max_heap_last_candidates = costs[max(candidates, len(costs) - candidates):]

        # Convert the lists into heaps
        heapify(min_heap_first_candidates)
        heapify(max_heap_last_candidates)

        total_cost = 0
        i = candidates
        j = len(costs) - candidates - 1

        # Perform k hiring sessions
        for _ in range(k):
            # Choose the worker with the lowest cost from either the first or last candidates
            if not max_heap_last_candidates or (min_heap_first_candidates and min_heap_first_candidates[0] <= max_heap_last_candidates[0]):
                total_cost += heappop(min_heap_first_candidates)
                if i <= j:
                    heappush(min_heap_first_candidates, costs[i])
                    i += 1
            else:
                total_cost += heappop(max_heap_last_candidates)
                if i <= j:
                    heappush(max_heap_last_candidates, costs[j])
                    j -= 1

        return total_cost
```

```python
# Example 1
costs = [17,12,10,2,7,2,11,20,8]
k = 3
candidates = 4
result = Solution().totalCost(costs, k, candidates)
print(result)
```

```python
# Example 2
costs = [1,2,4,1]
k = 3
candidates = 3
result = Solution().totalCost(costs, k, candidates)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


### Time Complexity:
- Initializing the heaps takes $O(candidates * log(candidates))$.
- Each iteration of the loop takes $O(log(candidates))$, and there are $k$ iterations.
- Hence, the overall time complexity is $O(candidates * log(candidates) + k * log(candidates))$.

### Space Complexity:
- The space complexity is $O(candidates)$ for the two heaps.
