## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use the two pointers technique.

Here's the Algorithm:

1. Initialize two pointers, left and right, both initially set to 0.

2. Iterate through the array using a loop:

  - If the element at `nums[left]` is not zero, move both left and right pointers to the next positions.

  - If the element at `nums[left]` is zero, move only the left pointer to the next position.

3. While the right pointer is less than the length of the array, set `nums[right]` to 0.

4. Return the modified nums array.

This Approach minimizes the total number of operations as we only swap elements when necessary, avoiding unnecessary swaps and additional operations.

By using the two pointers technique, we are able to move the zeros to the end of the array in-place while maintaining the relative order of the non-zero elements.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def moveZeroes(self, nums):
        left = 0
        right = 0

        while right < len(nums):
            if nums[left] != 0:
                left += 1
                right += 1
            elif nums[left] == 0 and nums[right] == 0:
                right += 1
            else:
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right += 1

        while right < len(nums):
            nums[right] = 0
            right += 1

        return nums
```

```python
#Example 1
nums = [0,1,0,3,12]
result = Solution().moveZeroes(nums)
print(result)
```

```python
#Example 2
nums = [0]
result = Solution().moveZeroes(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the solution is $O(n)$, where $n$ is the length of the input array `nums`. This is because we iterate through the array once using the two pointers.

The space complexity of the solution is $O(1)$ because we are modifying the input array in-place without using any additional data structures that grow with the input size. We do not allocate any extra memory proportional to the input size, resulting in constant space complexity.

Therefore, the time complexity is $O(n)$ and the space complexity is $O(1)$ for the given solution.
