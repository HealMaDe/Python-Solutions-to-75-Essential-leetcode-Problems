## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

1. Use a hash map to keep track of the frequency of each number in the array.

2. Iterate through the array and for each number num

  - Check if k - num exists in the hash map.
  - If it does, we increment the count of operations and decrement the frequency of both num and k - num in the hash map.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def maxOperations(self, nums, k):
        freq = {}
        count = 0

        for num in nums:
            if num in freq and freq[num] > 0:
                count += 1
                freq[num] -= 1
            else:
                complement = k - num
                freq[complement] = freq.get(complement, 0) + 1

        return count
```

```python
#Example 1
nums = [1,2,3,4]
k = 5
result = Solution().maxOperations(nums, k)
print(result)
```

```python
#Example 2
nums = [3,1,3,4,3]
k = 6
result = Solution().maxOperations(nums, k)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the length of the nums array, as we iterate through the array once.

The space complexity is also $O(n)$ because, in the worst-case scenario, all unique elements of the nums array will be stored in the hash map.
