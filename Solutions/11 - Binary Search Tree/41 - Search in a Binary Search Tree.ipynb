## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I utilize the properties of a binary search tree (BST)

1. search in the left subtree if the target value is less than the current node's value
2. search in the right subtree if the target value is greater
3. If the current node's value is equal to the target value, return the current node
4. If the target value is not found, return None.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def searchBST(self, root, val):
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode
        """
        # Base case: if the root is None or the value is found
        if not root or root.val == val:
            return root

        # If the value is less than the current root value, search in the left subtree
        if val < root.val:
            return self.searchBST(root.left, val)
        # If the value is greater than the current root value, search in the right subtree
        else:
            return self.searchBST(root.right, val)
```

```python
# Example 1
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
val = 2

def inorder_traversal(node):
    if not node:
        return []
    return inorder_traversal(node.left) + [node.val] + inorder_traversal(node.right)

result = Solution().searchBST(root,val)
print(inorder_traversal(result))
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(h)$, where $h$ is the height of the binary search tree (BST). In the worst case, the height of the BST could be equal to the number of nodes in the tree (n), leading to a time complexity of $O(n)$. However, in a balanced BST, the height is typically $log(n)$, resulting in an average time complexity of $O(log(n))$.

The space complexity is $O(h)$ as well, due to the recursive calls that are placed on the call stack. In the worst case, when the tree is skewed (essentially a linked list), the space complexity is $O(n)$. In a balanced BST, the space complexity is $O(log(n))$.
