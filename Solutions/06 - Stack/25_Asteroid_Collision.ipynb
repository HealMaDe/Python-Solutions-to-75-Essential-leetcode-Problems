## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a stack to simulate the movement of asteroids to the right. When a left-moving asteroid is encountered, it checks for collisions with the asteroids at the top of the stack and resolves them based on their sizes. If an asteroid survives the collisions, it is pushed onto the stack. The final state of the stack represents the result after all collisions.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def asteroidCollision(self, asteroids):
        stack = []

        for asteroid in asteroids:
            # Handle asteroids moving to the right
            if asteroid > 0:
                stack.append(asteroid)

            # Handle asteroids moving to the left
            else:
                while stack and stack[-1] > 0:
                    top_asteroid = stack.pop()
                    collision = abs(top_asteroid) - abs(asteroid)

                    if collision == 0:
                        # Both asteroids explode
                        break
                    elif collision < 0:
                        # The top asteroid explodes; continue checking with the next asteroid in the stack
                        continue
                    else:
                        # The new asteroid explodes; push it to the stack
                        stack.append(top_asteroid)
                        break
                else:
                    # No asteroid moving to the right to collide with the current left-moving asteroid
                    stack.append(asteroid)

        return stack
```

```python
#Example 1
asteroids = [5,10,-5]
result = Solution().asteroidCollision(asteroids)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png



### Time Complexity:
The time complexity is $O(n)$, where $n$ is the number of asteroids in the input array. In the worst case, each asteroid is processed exactly once. The while loop inside the for loop ensures that each asteroid is compared with the asteroids on the stack only once.

### Space Complexity:
The space complexity is also $O(n)$, where $n$ is the number of asteroids in the input array. In the worst case, all asteroids are moving to the right, and they will be stored in the stack. The stack size is proportional to the number of asteroids.
