## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a single pass through the string and use a stack to keep track of non-star indices. After processing all stars, we filter out the remaining non-star characters and stars from the result.

1. By using a a stack, iterate through the string once
2. for each character, check if it's a star or a non-star character
3. If it's a star, remove the closest non-star character to its left.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def removeStars(self, s):
        result = list(s)
        stack = []  # Stack to simulate non-star characters

        for i, char in enumerate(result):
            if char == '*':
                if stack:
                    # Remove the closest non-star character to the left
                    non_star_index = stack.pop()
                    result[non_star_index] = result[i] = '*'
            else:
                # Add the index of non-star characters to the stack
                stack.append(i)

        # Filter out the remaining non-star characters and stars
        result = [char for char in result if char != '*' and char != '-']

        return ''.join(result)
```

```python
#Example 1
s = "leet**cod*e"
result = Solution().removeStars(s)
print(result)
```

```python
#Example 2
s = "erase*****"
result = Solution().removeStars(s)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:

The time complexity is $O(n)$, where $n$ is the length of the input string. We iterate through the string once, and for each character, we perform constant-time operations. The use of a single pass through the string eliminates the nested loops present in the previous approach, making it more efficient.

<br>

Space Complexity:

The space complexity is also $O(n)$. We use a list (`result`) to store the characters of the string, and the size of this list is proportional to the length of the input string. Additionally, we use a stack to keep track of non-star indices, and in the worst case, the stack might contain all non-star indices. Therefore, the space complexity is linear with respect to the input size.
