## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a stack to keep track of counts and partial decoded strings.
1. iterate through the input string character by character
2. update the count and building the partial decoded string
3. Whenever you encounter brackets, push or pop elements from the stack accordingly.
4. The final result is the fully decoded string.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def decodeString(self, s):
        stack = []
        current_count = 0
        current_string = ""

        for char in s:
            if char.isdigit():
                current_count = current_count * 10 + int(char)
            elif char == '[':
                stack.append((current_count, current_string))
                current_count = 0
                current_string = ""
            elif char == ']':
                count, prev_string = stack.pop()
                current_string = prev_string + current_string * count
            else:
                current_string += char

        return current_string
```

```python
#Example 1
s = "3[a]2[bc]"
result = Solution().decodeString(s)
print(result)
```

```python
#Example 2
s = "3[a2[c]]"
result = Solution().decodeString(s)
print(result)
```

```python
#Example 3
s = "2[abc]3[cd]ef"
result = Solution().decodeString(s)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


### Time Complexity:
The time complexity is $O(n)$, where $n$ is the length of the input string. In the worst case, we iterate through each character in the string exactly once. Each character is processed in constant time, except when we encounter digits, where we may need to parse multi-digit numbers.

### Space Complexity:
The space complexity is also $O(n)$. In the worst case, the stack size is proportional to the length of the input string. The stack stores the counts and partial decoded strings, and its size is determined by the number of nested brackets in the input.
