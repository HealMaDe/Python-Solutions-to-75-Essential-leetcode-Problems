## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a depth-first search (DFS) traversal of the binary tree.

Here's a breakdown of the approach:

1. The `count_good_nodes` function takes a node and the maximum value encountered in the path so far (`max_val`).
2. If the current node is `None`, the function returns 0.
3. If the value of the current node is greater than or equal to `max_val`, it is considered a good node, and the count is incremented by 1.
4. The function then recursively calls itself on the left and right children of the current node, updating the `max_val` for each recursive call.
5. The total count of good nodes in the subtree rooted at the current node is the sum of the counts from the left and right recursive calls.
6. The function returns the total count.

This effectively explores the entire binary tree, updating the maximum value along the way to identify good nodes. The count is accumulated during the traversal, providing the final count of good nodes in the entire tree.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def goodNodes(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        def count_good_nodes(node, max_val=float('-inf')):
            if not node:
                return 0

            count = 0
            if node.val >= max_val:
                count += 1

            count += count_good_nodes(node.left, max(node.val, max_val))
            count += count_good_nodes(node.right, max(node.val, max_val))

            return count

        return count_good_nodes(root)
```

```python
# Example 1
root = TreeNode(3, TreeNode(1, TreeNode(3), TreeNode(1)), TreeNode(4, None, TreeNode(5)))
result = Solution().goodNodes(root)
print(result)
```

```python
#Example 2
root = TreeNode(3, TreeNode(3, None, TreeNode(4, None, TreeNode(2))))
result = Solution().goodNodes(root)
print(result)
```

```python
# Example 3:
root = TreeNode(1)
result = Solution().goodNodes(root)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the solution is $O(N)$, where $N$ is the number of nodes in the binary tree. This is because the solution uses a depth-first search (DFS) traversal of the entire tree, and each node is visited once.

The space complexity is $O(H)$, where $H$ is the height of the binary tree. This is due to the recursive nature of the solution. In the worst case, where the tree is skewed (essentially a linked list), the recursion depth can be equal to the height of the tree. In a balanced tree, the maximum recursion depth is $log(N)$. Therefore, the space complexity is dominated by the recursion depth. Additionally, the recursion stack contains information for each node on the path from the root to the current node.

In summary:
- Time Complexity: $O(N)$
- Space Complexity: $O(H)$
