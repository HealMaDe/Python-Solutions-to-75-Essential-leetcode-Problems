## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again, a recursive depth-first search (DFS) on the binary tree.

Here's a breakdown of the approach:

1. **Main Function (`count_paths`):**
   - The main function `count_paths` takes a `TreeNode` (`node`) and the target sum (`target`).
   - It starts by checking if the current node is `None` (base case). If so, it returns 0, as there are no paths from this point.
   - It then calls a helper function `count_paths_from_node` to count paths starting from the current node with the target sum.
   - Recursively, it counts paths in the left and right subtrees by calling itself on the left and right children of the current node.
   - The total number of paths is the sum of paths from the current node, paths from the left subtree, and paths from the right subtree.

2. **Helper Function (`count_paths_from_node`):**
   - The helper function `count_paths_from_node` takes a `TreeNode` (`node`) and the remaining target sum (`target`).
   - It starts by checking if the current node is `None`. If so, it returns 0.
   - It checks if the current node's value contributes to a path by comparing it to the remaining target. If so, it increments the count.
   - Recursively, it counts paths in the left and right subtrees by calling itself with the updated target (subtracting the current node's value).
   - The total number of paths is the sum of paths from the current node and paths from both subtrees.

The main function handles the overall counting of paths in the binary tree by combining the results from different nodes and subtrees. The recursive DFS explores the entire tree, considering each node as a potential starting point for a path.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def pathSum(self, root, targetSum):
        """
        :type root: TreeNode
        :type targetSum: int
        :rtype: int
        """
        def count_paths(node, target):
            if not node:
                return 0

            # Count paths starting from the current node
            paths_from_node = count_paths_from_node(node, target)

            # Recursively count paths in the left and right subtrees
            left_paths = count_paths(node.left, target)
            right_paths = count_paths(node.right, target)

            # Sum up the paths
            return paths_from_node + left_paths + right_paths

        def count_paths_from_node(node, target):
            if not node:
                return 0

            # Check if the current node contributes to a path
            paths = 1 if node.val == target else 0

            # Recursively count paths in the left and right subtrees
            paths += count_paths_from_node(node.left, target - node.val)
            paths += count_paths_from_node(node.right, target - node.val)

            return paths

        return count_paths(root, targetSum)
```

```python
# Example 1
root = TreeNode(10, TreeNode(5, TreeNode(3, TreeNode(3), TreeNode(-2)), TreeNode(2, None, TreeNode(1))), TreeNode(-3, None, TreeNode(11)))
targetSum = 8
result = Solution().pathSum(root,targetSum)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

**Time Complexity:**

The time complexity is $O(N)$, where $N$ is the number of nodes in the binary tree. This is because the algorithm visits each node once, and the work done at each node is constant.

**Space Complexity:**

The space complexity is $O(H)$, where $H$ is the height of the binary tree. In the worst case, the recursion stack would have a maximum depth equal to the height of the tree. In a balanced binary tree, the height is $log(N)$, but in the worst case (skewed tree), the height is $N$. The space complexity is dominated by the maximum depth of the recursion stack.
