## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a depth-first search (DFS) algorithm to explore the rooms and keep track of the visited rooms.

Here's the algorithm:

1. Initialize a set, `visited`, to keep track of visited rooms. Add room 0 to `visited`.
2. Create a stack and push room 0 onto the stack.
3. While the stack is not empty:
   - Pop a room, `currentRoom`, from the stack.
   - Iterate through the keys in `rooms[currentRoom]`:
     - If the key unlocks a room that has not been visited:
       - Add the key's corresponding room to `visited`.
       - Push the room onto the stack.
4. After the DFS traversal, if the number of visited rooms is equal to `n` (the total number of rooms), return `True`; otherwise, return `False`.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def canVisitAllRooms(self, rooms):
        n = len(rooms)
        visited = set()
        visited.add(0)
        stack = [0]

        while stack:
            currentRoom = stack.pop()
            for key in rooms[currentRoom]:
                if key not in visited:
                    visited.add(key)
                    stack.append(key)

        return len(visited) == n
```

```python
# Example 1
rooms = [[1],[2],[3],[]]
result = Solution().canVisitAllRooms(rooms)
print(result)
```

```python
# Example 2
rooms = [[1,3],[3,0,1],[2],[0]]
result = Solution().canVisitAllRooms(rooms)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the `canVisitAllRooms` function depends on the number of rooms (`n`) and the total number of keys across all rooms (`k`).

Time Complexity: $O(n + k)$
- In the worst case, we may need to visit all `n` rooms and process all `k` keys. Therefore, the time complexity is linearly proportional to the number of rooms and keys.

Space Complexity: $O(n)$
- We use the `visited` set to keep track of the visited rooms, which can have a maximum of `n` elements.
- Additionally, we use the stack to perform the DFS traversal. In the worst case, the stack may contain all `n` rooms if each room has a key that unlocks a new room.
- Therefore, the space complexity is linearly proportional to the number of rooms (`n`).
