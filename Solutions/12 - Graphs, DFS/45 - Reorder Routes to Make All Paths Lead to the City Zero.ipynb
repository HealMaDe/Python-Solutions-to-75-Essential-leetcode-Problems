## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again I use a Depth-First Search (DFS) approach to traverse the tree and count the number of edges that need to be changed.

Here is a breakdown of the Algorithm:

1. **Create an Adjacency List**: Convert the given connections into an adjacency list to represent the tree structure. Each entry in the list corresponds to a city, and the associated values are pairs (neighbor, is_forward), indicating the direction of the road (True for forward, False for backward).

2. **Depth-First Search (DFS)**: Implement a DFS function to traverse the tree starting from the capital city (city 0). During the DFS, count the number of forward edges encountered. The count represents the minimum number of edges that need to be changed to ensure that each city can reach the capital city.

3. **Recursive DFS Function**: The DFS function is a recursive function that explores the neighbors of each node, excluding the parent node. For each edge, it increments the count if the edge is a forward edge and recursively calls the function for the neighbor.

4. **Return the Count**: The final count is the minimum number of edges that need to be changed, and it is returned as the result.

The goal of this approach is to determine the direction in which each city should reorder its road connections to ensure that everyone can reach the capital city. The DFS explores the tree structure, keeping track of the count as it encounters forward edges.




## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def minReorder(self, n, connections):
        # Create an adjacency list to represent the tree
        graph = [[] for _ in range(n)]
        for connection in connections:
            graph[connection[0]].append((connection[1], True))  # True indicates a forward edge
            graph[connection[1]].append((connection[0], False))  # False indicates a backward edge

        # Perform Depth-First Search
        def dfs(node, parent, count):
            for neighbor, is_forward in graph[node]:
                if neighbor != parent:
                    count[0] += is_forward  # Increment count if the edge is a forward edge
                    dfs(neighbor, node, count)

        count = [0]  # Use a list to store the count
        dfs(0, -1, count)  # Start DFS from the capital city (node 0) with -1 as the parent
        return count[0]
```

```python
# Example 1
n = 6
connections = [[0,1],[1,3],[2,3],[4,0],[4,5]]
result = Solution().minReorder(n,connections)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

**Time Complexity:**

The time complexity is determined by the DFS traversal of the tree. In the worst case, each edge is visited once, as each city is visited exactly once. Therefore, the time complexity is $O(N)$, where $N$ is the number of cities.

**Space Complexity:**

1. **Graph Representation (Adjacency List):** $O(N + E)$, where $N$ is the number of cities, and $E$ is the number of edges. In the worst case, the graph is a straight line, and there are $N-1$ edges.

2. **DFS Recursive Call Stack:** $O(h)$, where $h$ is the height of the tree. In the worst case, if the tree is skewed, the height of the tree could be $N-1$, resulting in $O(N)$ space complexity for the call stack.

3. **Count List:** $O(1)$. The count list is a constant-size list.

Combining these, the overall space complexity is $O(N + h)$. In the worst case, if the tree is skewed, the space complexity becomes $O(N)$.
