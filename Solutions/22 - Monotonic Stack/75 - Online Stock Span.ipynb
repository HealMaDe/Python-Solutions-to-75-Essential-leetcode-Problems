## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I utilize a stack to keep track of the previous prices and their corresponding spans. I will store pairs of (price, span) in the stack, where the span represents the number of consecutive days with prices less than or equal to the current price.

Here's the Algorithm:

1. Initialize an empty stack to store the previous prices and their spans.

2. Take the current day's price as input and calculates the span for that day.

3.  Initialize the span set to 1 because there is at least one consecutive day with a price less than or equal to the current day's price (which is the current day itself).

4. heck the prices in the stack starting from the top.

5. If the price in the stack is less than or equal to the current day's price, add its span to the current span and remove it from the stack.
6. Repeat this process until we reach a price greater than the current day's price or the stack becomes empty.

7. Push the current day's price and its span to the stack.

8. Return the span for the current day.

The algorithm correctly calculates the spans for each day's stock price based on the previous prices, as shown in the example.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class StockSpanner(object):
    def __init__(self):
        self.stack = []

    def next(self, price):
        span = 1

        # Check the prices in the stack and add their spans
        while self.stack and self.stack[-1][0] <= price:
            span += self.stack.pop()[1]

        # Push the current price and its span to the stack
        self.stack.append((price, span))

        return span
```

```python
#Example 1
stockSpanner = StockSpanner()
print(stockSpanner.next(100))
print(stockSpanner.next(80))
print(stockSpanner.next(60))
print(stockSpanner.next(70))
print(stockSpanner.next(60))
print(stockSpanner.next(75))
print(stockSpanner.next(85))
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(k)$, where $k$ is the number of consecutive days with prices less than or equal to the current day's price. In the worst case, where the prices are strictly increasing, each price will be pushed and popped from the stack exactly once, resulting in a linear time complexity.

The space complexity is $O(n)$, where $n$ is the total number of calls. This is because the stack can store at most $n$ elements, representing all the previous prices and their spans.
