## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>
I utilize the Monotonic Decreasing Stack algorithm. We will traverse the array from left to right and maintain a decreasing monotonic stack of indices representing the days for which we haven't found a warmer temperature yet.

Here it goes:
1. Given the input array temperatures, initialize the result array with zeros and an empty stack.
2. Iterate through the temperatures from left to right.
  - For each temperature, compare it with the temperature at the top of the stack.

  - If the temperature at the top of the stack is lower than temperature, it means we have found a warmer temperature for the day represented by the index at the top of the stack.
  - pop the index from the stack
  - calculate the number of days waited
  - update the corresponding entry in the result array.

  - If the temperature at the top of the stack is higher than or equal to temperatures, continue to the next day by pushing the current index onto the stack.

3. Return the result array, which contains the number of days waited to get a warmer temperature for each day. If there is no future day with a warmer temperature, the corresponding entry in the result array remains zero.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def dailyTemperatures(self, temperatures):

        n = len(temperatures)
        result = [0] * n
        stack = []

        for i in range(n):
            while stack and temperatures[stack[-1]] < temperatures[i]:
                index = stack.pop()
                result[index] = i - index
            stack.append(i)

        return result
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the `dailyTemperatures` algorithm is $O(n)$, where $n$ is the length of the input array `temperatures`. This is because we traverse the array once, and for each element, we perform constant time operations (pushing, popping, and updating the result array) while maintaining the monotonic decreasing stack. Therefore, the algorithm has a linear time complexity.

The space complexity of the algorithm is $O(n)$, where $n$ is the length of the input array `temperatures`. This is because in the worst case scenario, where temperatures are in non-decreasing order, the stack will contain all n elements. Additionally, the result array has a length of $n$ to store the number of days waited for each day. Hence, the space complexity is linear with respect to the input size.
