## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again, I use a sliding window approach.

Here's a breakdown of the algorithm:

1. Initialize variables: max_length to keep track of the maximum length of the subarray containing only 1's, left as the left pointer of the sliding window, and count_zeros to count the number of zeros in the current window.

2. Iterate through the array using the right pointer (right). For each element:

  - If the element is 0, increment count_zeros.
  - Check if there are more than one zero in the current window. If yes, shrink the window by incrementing the left pointer until there is at most one zero in the window.
3. Update the max_length by taking the maximum of the current max_length and the length of the current subarray (right - left + 1).

4. Return max_length - 1 as the result, as one element needs to be deleted to maximize the length of the subarray.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def longestSubarray(self, nums):
        max_length = 0  # Variable to store the maximum length of subarray
        left = 0        # Left pointer of the sliding window
        count_zeros = 0 # Count of zeros in the current window

        for right in range(len(nums)):
            if nums[right] == 0:
                count_zeros += 1

            # Shrink the window if there are more than one zero in the window
            while count_zeros > 1:
                if nums[left] == 0:
                    count_zeros -= 1
                left += 1

            # Update the maximum length of the subarray
            max_length = max(max_length, right - left + 1)

        return max_length - 1  # Return the length after deleting one element
```

```python
#Example 1
nums = [1,1,0,1]
result = Solution().longestSubarray(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the provided code is $O(n)$, where n is the length of the input array nums. This is because the algorithm iterates through the array once using a sliding window approach, and each element is processed at most twice (once when it enters the window and once when it leaves the window).

The space complexity is $O(1)$ because the algorithm uses a constant amount of extra space regardless of the size of the input array. The only additional variables used are max_length, left, and count_zeros, and they do not depend on the size of the input array. The space required to store these variables remains constant as the input size increases.
