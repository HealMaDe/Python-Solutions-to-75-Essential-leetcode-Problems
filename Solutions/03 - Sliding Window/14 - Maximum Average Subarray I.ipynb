## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use the sliding window technique.

1. Maintain a window of size k and slide it through the array
2. Calculatethe sum of the current window
3. Update the maximum average as you go.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def findMaxAverage(self, nums, k):
        window_sum = sum(nums[:k]) # Initialize the sum of the first window
        max_average = window_sum

        for i in range(k, len(nums)):
            window_sum += nums[i] - nums[i - k] # Update the window sum by adding the next element and removing the first element of the window
            max_average = max(max_average, window_sum )


        return max_average/(k*1.00)
```

```python
#Example 1
nums = [1,12,-5,-6,50,3]
k = 4
result = Solution().findMaxAverage(nums, k)
print(result)
```

```python
#Example 2
nums = [5]
k = 1
result = Solution().findMaxAverage(nums, k)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the length of the nums array. We iterate through the array once while maintaining the sliding window.

The space complexity is $O(1)$ since we only use a constant amount of extra space to store variables.
