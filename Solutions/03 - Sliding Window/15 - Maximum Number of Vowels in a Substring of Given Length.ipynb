## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again, I use a sliding window approach.

1. Maintain a window of length k and slide it through the string
2. Count the number of vowel letters in each window
3. Keep track of the maximum count.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def maxVowels(self, s, k):
        vowels = set(['a', 'e', 'i', 'o', 'u'])
        max_count = 0
        current_count = 0

        # Count the vowels in the first window
        for i in range(k):
            if s[i] in vowels:
                current_count += 1

        max_count = current_count

        # Slide the window and update the counts
        for i in range(k, len(s)):
            if s[i - k] in vowels:
                current_count -= 1
            if s[i] in vowels:
                current_count += 1
            max_count = max(max_count, current_count)

        return max_count
```

```python
#Example 1
s = "abciiidef"
k = 3
result = Solution().maxVowels(s, k)
print(result)
```

```python
#Example 2
s = "aeiou"
k = 2
result = Solution().maxVowels(s, k)
print(result)
```

```python
#Example 3
s = "leetcode"
k = 3
result = Solution().maxVowels(s, k)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the length of the input string $s$. This is because we iterate through the string twice: once to count the vowels in the first window of length $k$, and then again to slide the window and update the counts.

The space complexity is $O(1)$ because we only use a constant amount of additional space to store the vowels set and a few variables to keep track of the maximum count and current count.
