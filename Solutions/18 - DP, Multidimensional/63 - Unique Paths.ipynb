## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

multidimensional DP. I create a 2D array dp of size m x n to store the number of unique paths for each position on the grid.

1. initialize the dp array with all elements set to 0
2. iterate through each position on the grid using two nested loops.
3. For the positions in the first row and the first column, set the value to 1 since there is only one path to reach those positions (either moving right or moving down).
4. For the other positions, calculate the number of unique paths based on the recurrence relation.
5. return the value at the bottom-right corner of the grid, which represents the number of unique paths to reach the destination.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def uniquePaths(self, m, n):
        dp = [[0] * n for _ in range(m)]

        for i in range(m):
            for j in range(n):
                if i == 0 or j == 0:
                    dp[i][j] = 1
                else:
                    dp[i][j] = dp[i-1][j] + dp[i][j-1]

        return dp[m-1][n-1]
```

```python
# Example 1
m = 3
n = 7
result = Solution().uniquePaths(m,n)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(m * n)$, where $m$ and $n$ are the dimensions of the grid.

The space complexity is also $O(m * n)$ since we create a 2D array to store the intermediate results.
