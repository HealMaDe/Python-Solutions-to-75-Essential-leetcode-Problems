## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

multidimensional DP.

maintain two states: hold and not_hold, representing the maximum profit we can achieve if we hold a stock or if we don't hold a stock, respectively.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def maxProfit(self, prices, fee):
        n = len(prices)
        hold = [0] * n
        not_hold = [0] * n

        hold[0] = -prices[0]

        for i in range(1, n):
            hold[i] = max(hold[i-1], not_hold[i-1] - prices[i])
            not_hold[i] = max(not_hold[i-1], hold[i-1] + prices[i] - fee)

        return not_hold[n-1]
```

```python
# Example 1
prices = [1,3,2,8,4,9]
fee = 2
result = Solution().maxProfit(prices, fee)
print(result)
```

```python
# Example 2
prices = [1,3,7,5,10,3]
fee = 3
result = Solution().maxProfit(prices, fee)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the length of the prices array.

The space complexity is also $O(n)$ since we create two arrays to store the intermediate results.
