## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use dynamic programming to calculate the Tribonacci sequence up to the given value of n. create a memoization array to store the values of the Tribonacci sequence as calculating them.




## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def tribonacci(self, n):
        if n == 0:
            return 0
        elif n <= 2:
            return 1

        dp = [0] * (n + 1)
        dp[0] = 0
        dp[1] = 1
        dp[2] = 1

        for i in range(3, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2] + dp[i - 3]

        return dp[n]
```

```python
# Example 1
n = 4
result = Solution().tribonacci(n)
print(result)
```

```python
# Example 2
n = 25
result = Solution().tribonacci(n)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(n)$. This is because we iterate $n$ times to fill up the memoization array, where $n$ represents the given value for which we want to compute the Tribonacci number.

The space complexity of the solution is also $O(n)$. We create a memoization array, `dp`, of size n+1 to store the values of the Tribonacci sequence. Therefore, the space required is proportional to the value of $n$.

Since the given constraints guarantee that $n <= 37$, both the time and space complexities are considered efficient for practical purposes.
