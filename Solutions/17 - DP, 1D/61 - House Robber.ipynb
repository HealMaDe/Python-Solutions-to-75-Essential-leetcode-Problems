## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again, Another DP approach:

1. handle the base cases where the nums array has one or two elements and return the corresponding maximum amount directly.
2. initialize the dp array with the base cases.
3. iterate from the 2nd index to the last index of the nums array, calculating the maximum amount of money that can be robbed up to each house and storing it in the dp array.
4. return the maximum amount among the last two elements of the dp array, as the maximum can be achieved by either robbing the last house or skipping it.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def rob(self, nums):
        n = len(nums)
        if n == 1:
            return nums[0]
        elif n == 2:
            return max(nums[0], nums[1])

        dp = [0] * n
        dp[0] = nums[0]
        dp[1] = max(nums[0], nums[1])

        for i in range(2, n):
            dp[i] = max(dp[i-1], dp[i-2] + nums[i])

        return dp[n-1]
```

```python
# Example 1
nums = [1,2,3,1]
result = Solution().rob(nums)
print(result)
```

```python
# Example 2
nums = [2,7,9,3,1]
result = Solution().rob(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the length of the nums array, as we iterate through the array once.

The space complexity is $O(n)$ as well, since we create an array of the same size as the nums array to store the intermediate results.
