## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Again, Another DP approach:

1.  initialize the dp array with base cases
2. iterate from 3 to n, calculating the number of ways to tile a 2 x i board based on the recurrence relation.
3. return dp[n], which represents the number of ways to tile a 2 x n board.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def numTilings(self, n):
        MOD = 10**9 + 7
        dp = [0] * (n + 1)
        dp[0] = 1
        dp[1] = 1

        if n > 1:
            dp[2] = 2

        for i in range(3, n + 1):
            dp[i] = (dp[i-1] + dp[i-2] + 2 * sum(dp[:i-2])) % MOD

        return dp[n]
```

```python
# Example 1
n = 3
result = Solution().numTilings(n)
print(result)
```

```python
# Example 2
n = 1
result = Solution().numTilings(n)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(n)$, where $n$ is the given value.

The space complexity is $O(n)$ as well, since we create an array of the same size to store the intermediate results.
