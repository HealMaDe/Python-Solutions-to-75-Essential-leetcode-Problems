## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a hash map to keep track of the number of occurrences of each value in the array. If the counts of occurrences are unique, then the answer is true; otherwise, it is false.




## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def uniqueOccurrences(self, arr):
        count_map = {}

        # Count the occurrences of each value in the array
        for num in arr:
            if num in count_map:
                count_map[num] += 1
            else:
                count_map[num] = 1

        # Check if the counts of occurrences are unique
        occurrences = count_map.values()
        return len(occurrences) == len(set(occurrences))
```

```python
#Example 1
arr = [1,2,2,1,1,3]
result = Solution().uniqueOccurrences(arr)
print(result)
```

```python
#Example 2
arr = [1,2]
result = Solution().uniqueOccurrences(arr)
print(result)
```

```python
#Example 3
arr = [-3,0,1,-3,1,1,1,-3,10,0]
result = Solution().uniqueOccurrences(arr)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- The loop that counts the occurrences of each value in the array has a time complexity of $O(n)$, where $n$ is the length of the input array `arr`. This is because we iterate over each element in the array once.
- Creating the set of occurrences using the `set()` function has a time complexity of $O(k)$, where $k$ is the number of unique occurrences in the array.
- Therefore, the overall time complexity of the code is $O(n + k)$.

Space Complexity:
- The space complexity of the code depends on two main factors: the hash map `count_map` and the set of occurrences.
- The hash map `count_map` stores the occurrences of each value in the array. In the worst case, if all elements in the array are unique, the size of the hash map can be equal to the length of the array, resulting in a space complexity of $O(n)$.
- The set of occurrences is created using the `set()` function, which takes additional space proportional to the number of unique occurrences. In the worst case, if all elements in the array have distinct occurrences, the size of the set will be equal to the length of the array, resulting in a space complexity of $O(n)$.
- Therefore, the overall space complexity of the code is O(n) in the worst case.

In summary, the code has a time complexity of $O(n + k)$ and a space complexity of $O(n)$, where $n$ is the length of the input array `arr` and $k$ is the number of unique occurrences in the array.
