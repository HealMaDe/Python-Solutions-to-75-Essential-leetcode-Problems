## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

By using hash maps to store the rows and columns, I achieve an efficient approach.

Here it goes:

1. Initialize a variable to keep track of the number of equal row and column pairs.

2. Initialize an empty dictionary called to store the rows encountered.

3. Iterate over each row from 0 to n-1.

   - Convert the current row into a tuple and add it to the dictionary. The key will be the tuple representing the row, and the value will be the count of occurrences.

4. Iterate over each column from 0 to n-1.

   - Initialize an empty list called to store the elements in the current column.

   - Iterate over each row from 0 to n-1.

     - Append the element to the list.

   - Convert the list into a tuple.

   - If the column tuple is present in the dictionary, increment by the count of occurrences for that row.

5. Return the value of count, which represents the total number of equal row and column pairs.

The time complexity of this Approach is $O(n^2)$ since we need to iterate over each element in the $n$ x $n$ grid. However, by using hash maps, we can achieve a more efficient lookup when checking for the presence of a column tuple in the rows dictionary.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def equalPairs(self, grid):
        n = len(grid)
        count = 0
        rows = {}

        for ri in range(n):
            row_tuple = tuple(grid[ri])
            rows[row_tuple] = rows.get(row_tuple, 0) + 1

        for cj in range(n):
            column = []
            for ri in range(n):
                column.append(grid[ri][cj])
            column_tuple = tuple(column)
            if column_tuple in rows:
                count += rows[column_tuple]

        return count
```

```python
#Example 1
grid = [[3,2,1],[1,7,6],[2,7,7]]
result = Solution().equalPairs(grid)
print(result)
```

```python
#Example 2
grid = [[3,1,2,2],[1,4,4,5],[2,4,2,2],[2,4,2,2]]
result = Solution().equalPairs(grid)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Constructing the `rows` dictionary takes $O(n^2)$ time since we iterate over each element in the $n$ x $n$ grid.
- Constructing the `column` list for each column also takes $O(n^2)$ time since we iterate over each element in the grid for each column.
- Checking for the presence of the column tuple in the `rows` dictionary takes $O(1)$ time on average due to the efficient lookup in hash maps.
- Therefore, the overall time complexity of the solution is $O(n^2)$.

Space Complexity:
- The space complexity is determined by the `rows` dictionary, which stores the rows encountered.
- In the worst case, if all rows are unique, the `rows` dictionary will have a size of $n$ x $n$ to store all unique rows.
- Therefore, the space complexity of the solution is $O(n^2)$.
