## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

Here how it goes:

1. Convert `nums1` and `nums2` to sets, `s1` and `s2`, respectively, using the `set()` function. This eliminates any duplicate elements and improves lookup efficiency.
2. Use the `difference()` method of sets to find the distinct integers that are present in one set but not in the other.
3. Convert the resulting sets back to lists using the `list()` function.
4. Return the lists as the answer.




## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def findDifference(self, nums1, nums2):
        s1 , s2 = set(nums1) , set(nums2)
        result1 = s1.difference(s2)
        result2 = s2.difference(s1)
        return [list(result1),list(result2)]
```

```python
#Example 1
nums1 = [1,2,3]
nums2 = [2,4,6]
result = Solution().findDifference(nums1, nums2)
print(result)
```

```python
#Example 2
nums1 = [1,2,3,3]
nums2 = [1,1,2,2]
result = Solution().findDifference(nums1, nums2)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:
1. Converting `nums1` and `nums2` to sets takes $O(n)$ time, where $n$ is the length of the input lists.
2. The `difference()` method on sets has an average time complexity of $O(min(len(s1), len(s2)))$ in Python.
3. Converting the resulting sets back to lists takes $O(k)$ time, where $k$ is the size of the resulting sets.
4. Therefore, the overall time complexity of the code is $O(n + k)$, where $n$ is the size of the input lists and $k$ is the size of the resulting sets.
5. Since the code performs a linear scan of the input lists and the size of the resulting sets is limited by the number of distinct integers, the code runs in linear time complexity.

Space Complexity:
1. The code uses additional space to store the sets `s1` and `s2`, which have a space complexity of $O(n)$, where $n$ is the size of the input lists.
2. The resulting sets `result1` and `result2` are converted to lists, which have a space complexity of $O(k)$, where $k$ is the size of the resulting sets.
3. Therefore, the overall space complexity of the code is $O(n + k)$, where $n$ is the size of the input lists and $k$ is the size of the resulting sets.
