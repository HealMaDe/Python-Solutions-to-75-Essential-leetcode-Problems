## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use dictionaries to count the frequencies of characters. This approach avoids sorting the frequency lists, resulting in a more efficient solution.

Algorithm:

1. Check if the lengths of word1 and word2 are equal, returning False if they are not.

2. Use two dictionaries to count the frequencies of characters in word1 and word2, respectively.

3. Iterate over each character in word1 and update its frequency to handle both new and existing characters. We do the same for word2.

4. Check if the two strings have the same set of characters by comparing the sets of keys in dictionaries.

5. Extract the frequency values from dictionaries, sort them and compare them.


## 5. Code Implementation <a name="Code-Implementation"></a>


```python

class Solution(object):
    def closeStrings(self, word1, word2):
        if len(word1) != len(word2):
            return False

        # Count the frequencies of characters in word1
        freq1 = {}
        for ch in word1:
            freq1[ch] = freq1.get(ch, 0) + 1

        # Count the frequencies of characters in word2
        freq2 = {}
        for ch in word2:
            freq2[ch] = freq2.get(ch, 0) + 1

        # Check if the two strings have the same set of characters
        if set(freq1.keys()) != set(freq2.keys()):
            return False

        # Check if the two strings have the same frequencies for each character
        freq1_values = sorted(freq1.values())
        freq2_values = sorted(freq2.values())

        return freq1_values == freq2_values
```

```python
#Example 1
word1 = "abc"
word2 = "bca"
result = Solution().closeStrings(word1, word2)
print(result)
```

```python
#Example 2
word1 = "a"
word2 = "aa"
result = Solution().closeStrings(word1, word2)
print(result)
```

```python
#Example 3
word1 = "cabbba"
word2 = "abbccc"
result = Solution().closeStrings(word1, word2)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Counting the frequencies of characters in both `word1` and `word2` takes $O(n)$, where $n$ is the length of the input strings. This is because we iterate over each character in the strings once.
- Checking if the two strings have the same set of characters involves comparing the sets of keys in the dictionaries. This operation has a time complexity of $O(k)$, where $k$ is the number of unique characters in the input strings. In the worst case, where all characters are unique, $k$ can be at most $min(n, m)$, where $n$ and $m$ are the lengths of the input strings. Therefore, the overall time complexity for this step is $O(min(n, m))$.
- Extracting the frequency values from the dictionaries and sorting them has a time complexity of $O(k *log (k))$ in the worst case, where $k$ is the number of unique characters. Again, in the worst case, $k$ can be at most $min(n, m)$, so the overall time complexity for this step is $O(min(n, m) *log (min(n, m)))$.

Therefore, the overall time complexity of the code is $O(n + min(n, m) + min(n, m)* log (min(n, m)))$, which simplifies to $O(n + min(n, m) *log (min(n, m)))$.

Space Complexity:
- The space complexity is determined by the dictionaries used to count the frequencies of characters. In the worst case, when all characters are unique, the dictionaries will have a size of $min(n, m)$ to store the frequencies of all unique characters.
- Therefore, the space complexity of the code is $O(min(n, m))$.
