## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use the greedy approach. The idea is to sort the intervals based on their end points and then iterate through the sorted intervals, keeping track of the current end point. If the start point of the current interval is less than the current end point, it means there is an overlap and we need to remove one interval. Otherwise, we update the current end point to the end point of the current interval.

Here's the step-by-step algorithm:

1. Sort the intervals based on their end points.
2. Initialize a variable count to 0 to keep track of the number of intervals to be removed.
3. Initialize a variable end to the end point of the first interval in the sorted list.
4. Iterate through the remaining intervals starting from the second interval.
5. If the start point of the current interval is less than the current end, it means there is an overlap:
  - Increment count by 1.
  - Update end to the minimum of end and the end point of the current interval.
6. Otherwise, update end to the end point of the current interval.
7. After the loop ends, return the value of count.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def eraseOverlapIntervals(self, intervals):
        intervals.sort(key=lambda x: x[1])  # Sort intervals based on their end points
        count = 0
        current_end = intervals[0][1]

        for i in range(1, len(intervals)):
            if intervals[i][0] < current_end:
                count += 1
            else:
                current_end = intervals[i][1]

        return count
```

```python
#Example 1
intervals = [[1,2],[2,3],[3,4],[1,3]]
result = Solution().eraseOverlapIntervals(intervals)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

Time Complexity:
- Sorting the intervals takes $O(n *log (n))$, where $n$ is the number of intervals.
- The for loop iterates through the intervals once, taking $O(n)$ time.
- Therefore, the overall time complexity is dominated by the sorting operation, resulting in $O(n *log (n))$.

Space Complexity:
- The code uses a constant amount of extra space for variables such as `count` and `current_end`, so they contribute $O(1)$ space.
- The sorting operation is performed in-place on the `intervals` list, so it does not require any additional space.
- Hence, the space complexity is $O(1)$.
