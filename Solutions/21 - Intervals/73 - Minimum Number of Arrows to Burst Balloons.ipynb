## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>
I use the interval scheduling algorithm, similar to the previous problem. The idea is to sort the intervals based on their end points and then count the number of non-overlapping intervals.

Here it goes:

1. Sort the intervals based on their end points in non-decreasing order.
2. Initialize a variable count to 1 to keep track of the number of arrows needed.
3. Initialize a variable end to the end point of the first interval in the sorted list.
4. Iterate through the remaining intervals:
  - If the start point of the current interval is greater than end, it means the current interval is not overlapping with the previous ones. Increment count by 1 and update end to the end point of the current interval.
5. After the loop ends, count will contain the minimum number of arrows needed.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def findMinArrowShots(self, points):
        points.sort(key=lambda x: x[1])  # Sort intervals based on their end points
        count = 1
        current_end = points[0][1]

        for i in range(1, len(points)):
            if points[i][0] > current_end:
                count += 1
                current_end = points[i][1]

        return count
```

```python
#Example 1
points = [[10,16],[2,8],[1,6],[7,12]]
result = Solution().findMinArrowShots(points)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


Time Complexity:
- Sorting the points takes $O(n*log (n))$ time, where $n$ is the number of intervals.
- The for loop iterates through the points once, taking $O(n)$ time.
- Therefore, the overall time complexity is dominated by the sorting operation, resulting in $O(n*log (n))$.

Space Complexity:
- The code uses a constant amount of extra space for variables such as `count` and `current_end`, so they contribute $O(1)$ space.
- The sorting operation is performed in-place on the `points` list, so it does not require any additional space.
- Hence, the space complexity is $O(1)$.
