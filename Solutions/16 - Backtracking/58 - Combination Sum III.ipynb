## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use use a backtracking approach to generate all possible combinations of k numbers that sum up to n

Define a backtrack function that explores all possible combinations starting from a given position (start). It chooses each number, explores further, and then unchooses (backtracks) to try other possibilities.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def combinationSum3(self, k, n):
        def backtrack(start, k, n, path, result):
            # Base case: if k becomes 0 and n is 0, add the current combination to the result
            if k == 0 and n == 0:
                result.append(path[:])
                return

            # Explore possibilities
            for i in range(start, 10):
                # Choose
                path.append(i)

                # Explore
                backtrack(i + 1, k - 1, n - i, path, result)

                # Unchoose (backtrack)
                path.pop()

        result = []
        backtrack(1, k, n, [], result)
        return result
```

```python
#Example 1
k = 3
n = 7
result = Solution().combinationSum3(k, n)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time Complexity:

The time complexity of the code is exponential, as the algorithm explores all possible combinations through recursion. The number of combinations grows with the input values `k` and `n`. In the worst case, the code needs to explore all combinations of size `k` from the set of numbers 1 through 9.

Considering that the depth of the recursion is `k` and, at each level, we have at most 9 choices (numbers 1 through 9), the time complexity is approximately $O(9^k)$. However, the actual number of combinations may be smaller, as the algorithm may terminate early in some branches.

<br>

Space Complexity:

The space complexity is determined by the recursion depth and the space required to store combinations. The recursion depth is at most `k`, and at each level of recursion, a combination of size `k` is stored temporarily in the `path` list. Therefore, the space complexity is $O(k)$ for the recursion depth and $O(k)$ for each recursive call, resulting in a total space complexity of $O(k^2)$.

In addition to the recursive call stack, the `result` list is used to store the valid combinations. The space required for the `result` list is proportional to the number of valid combinations generated, which can vary but is still within the bounds of the constraints.
