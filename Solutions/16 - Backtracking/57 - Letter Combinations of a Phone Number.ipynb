## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a recursive approach to generate all possible combinations.

1. Define a that function uses a backtracking approach to generate all possible combinations of letters corresponding to the input digits.
2. The backtrack function is called recursively for each digit, and the combinations are built up gradually.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def letterCombinations(self, digits):
        if not digits:
            return []

        # Define a mapping of digits to letters
        digit_to_letters = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz'
        }

        result = []

        def backtrack(index, path):
            # If the current path is of the same length as digits, add it to the result
            if index == len(digits):
                result.append(''.join(path))
                return

            # Get the letters corresponding to the current digit
            current_letters = digit_to_letters[digits[index]]

            # Iterate through the letters and recursively explore possibilities
            for letter in current_letters:
                path.append(letter)
                backtrack(index + 1, path)
                path.pop()

        backtrack(0, [])
        return result
```

```python
#Example 1
digits = "23"
result = Solution().letterCombinations(digits)
print(result)
```

```python
#Example 2
digits = ""
result = Solution().letterCombinations(digits)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of this solution is $O(4^n)$, where $n$ is the length of the input digits. In the worst case, each digit has four corresponding letters, and there are $n$ digits.

The space complexity is $O(n)$ due to the recursive call stack.

**Note:** While the time complexity is exponential in the worst case, the actual number of combinations is not large (due to the limited number of digits and the limited number of letters associated with each digit).
