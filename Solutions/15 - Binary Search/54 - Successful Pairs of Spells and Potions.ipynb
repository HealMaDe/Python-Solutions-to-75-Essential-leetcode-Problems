## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I utilize binary search

1. Sort the potions array to ensure that binary search can be applied.
2. For each element in spells array, perform a binary search to find the position where spell * potion strength is greater than or equal to success.
3. The count of successful pairs is then calculated as the difference between the total number of potions and the position found in the binary search.
4. Repeat this process for each spell and store the counts in the pairs array.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
import bisect
class Solution(object):
    def successfulPairs(self, spells, potions, success):
        # Sort the potions array
        potions.sort()

        # Initialize an array to store the count of successful pairs for each spell
        pairs = [0] * len(spells)

        # Iterate through each spell
        for i in range(len(spells)):
            spell = spells[i]
            left = 0
            right = len(potions) - 1

            # Perform binary search to find the position where spell * potion strength is >= success
            while left <= right:
                mid = left + (right - left)//2
                if spell * potions[mid] >= success:
                    right = mid - 1
                else:
                    left = mid + 1

            # The count of successful pairs is given by the difference between the total potions
            # and the position found in the binary search
            pairs[i] = len(potions) - left

        # Return the array containing the count of successful pairs for each spell
        return pairs
```

```python
# Example 1
spells = [5,1,3]
potions = [1,2,3,4,5]
success = 7
result = Solution().successfulPairs(spells, potions, success)
print(result)
```

```python
# Example 2
spells = [3,1,2]
potions = [8,5,8]
success = 16
result = Solution().successfulPairs(spells, potions, success)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

Time Complexity:
The time complexity is $O(n *log (m))$, where $n$ is the length of the spells array and $m$ is the length of the potions array. The dominant factor is the binary search within the loop.

Space Complexity:
The space complexity is $O(n)$, where $n$ is the length of the spells array. The additional space is used to store the pairs array.
