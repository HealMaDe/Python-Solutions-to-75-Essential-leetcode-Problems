## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a modified form of binary search. The idea is to compare the middle element with its neighbors to determine if it is a peak. If the middle element is not a peak, we move towards the direction of the higher neighbor, as there must be a peak in that direction.

Here's the algorithm to find a peak element:

1. Initialize two pointers, `left` and `right`, pointing to the start and end of the array, respectively.
2. While `left < right`, do the following:
   - Calculate the middle index as `mid = left + (right - left) // 2`.
   - Compare `nums[mid]` with its neighbors `nums[mid - 1]` and `nums[mid + 1]`.
     - If `nums[mid]` is greater than both neighbors, return `mid` as the peak index.
     - If `nums[mid]` is smaller than `nums[mid + 1]`, move `left` to `mid + 1` to search in the higher part of the array.
     - Otherwise, move `right` to `mid - 1` to search in the lower part of the array.
3. If the loop exits without finding a peak, it means the peak is either at index `left` or `right`. Return either index as the peak index.

## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def findPeakElement(self, nums):
        left = 0
        right = len(nums) - 1

        while left < right:
            mid = left + (right - left) // 2

            if nums[mid] > nums[mid + 1]:
                right = mid
            else:
                left = mid + 1

        return left
```

```python
# Example 1
nums = [1,2,3,1]
result = Solution().findPeakElement(nums)
print(result)
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of the algorithm is $O(log (n)), where $n$ is the length of the input array `nums`. This is because the algorithm uses binary search to narrow down the search space in half at each step. As a result, the number of iterations required to find the peak element is proportional to the logarithm of the input size.

The space complexity of the algorithm is $O(1)$ because it uses a constant amount of extra space to store the pointers `left` and `right`. The space usage does not depend on the size of the input array.
