## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I utilize binary search to efficiently guess the picked number.

Here's an overview of the approach:
1. Initialize the left pointer (left) to 1 and the right pointer (right) to n, representing the range of possible numbers.
2. While the left pointer is less than or equal to the right pointer:
  - Calculate the middle index as mid = left + (right - left) // 2.
  - Use the guess function to determine the relationship between the mid value and the picked number.
  - If the guess is correct (result is 0), return mid as the picked number.
  - If the guess is lower (result is 1), update the right pointer to mid - 1 to search in the lower half.
  - If the guess is higher (result is -1), update the left pointer to mid + 1 to search in the upper half.
3. If the loop terminates without finding the picked number, return -1 to indicate that the number was not found.



## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def guessNumber(self, n):
        left = 1
        right = n

        while left <= right:
            mid = left + (right - left) // 2
            result = guess(mid)

            if result == 0:
                return mid
            elif result == -1:
                right = mid - 1
            else:
                left = mid + 1

        return -1  # Number not found

def guess(num):
    if num < pick:
        return 1
    elif num > pick:
        return -1
    else:
        return 0
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(log (n))$ because it uses a binary search algorithm to find the picked number. In each iteration of the binary search loop, the search space is halved, which leads to a logarithmic time complexity.

The space complexity of the given code is $O(1)$ because it uses a constant amount of extra space. It only uses a few variables to keep track of the left and right pointers, the middle index, and the result of the guess function. The space used by these variables does not depend on the input size.
