## 4. Approach and Algorithm <a name="Approach-and-Algorithm"></a>

I use a binary search approach.

Here's the algorithm to find the minimum eating speed:

1. Initialize the lower bound `left` as 1 and the upper bound `right` as the maximum number of bananas in the `piles` array.
2. While `left` is less than or equal to `right`, do the following:
   - Calculate the middle speed as `mid = left + (right - left) // 2`.
   - Simulate Koko's eating process by traversing through the `piles` array and calculating the total hours required to eat all the bananas.
     - For each pile `p` in `piles`, calculate the hours required to eat the bananas in that pile as `hours = (p + mid - 1) // mid`. Add `hours` to the total.
   - Compare the total hours with `h`.
     - If the total hours is greater than `h`, it means Koko needs to eat at a higher speed. So, update `left = mid + 1`.
     - Otherwise, the total hours is less than or equal to `h`, indicating that Koko can finish eating within `h` hours. Update `right = mid - 1`.
3. After the loop ends, the minimum eating speed will be in the `left` variable. Return `left` as the minimum eating speed.




## 5. Code Implementation <a name="Code-Implementation"></a>


```python
class Solution(object):
    def minEatingSpeed(self, piles, h):
        left = 1
        right = max(piles)

        while left <= right:
            mid = left + (right - left) // 2
            total_hours = 0

            for p in piles:
                total_hours += (p + mid - 1) // mid

            if total_hours > h:
                left = mid + 1
            else:
                right = mid - 1

        return left
```

## 6. Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity is $O(n *log (m))$, where $n$ is the number of piles and m is the maximum number of bananas in a pile. This is because the algorithm uses binary search to find the minimum eating speed, and in each iteration of the binary search, it calculates the total hours by traversing through all the piles. The binary search has a time complexity of $O(log (m))$, and in each iteration, the algorithm takes $O(n)$ time to calculate the total hours. Therefore, the overall time complexity is $O(n *log (m))$.

The space complexity of the algorithm is $O(1)$ because it uses a constant amount of extra space to store the lower and upper bounds, as well as the variables for the binary search and total hours calculations. The space usage does not depend on the size of the input.
